# **Modules in Python**
## **1. What is a Module?**
A **module** in Python is a file containing Python **functions, classes, and variables** that can be reused across multiple programs. It helps in **code reusability and organization**.

- A module is simply a **Python file** with a **`.py`** extension.
- It can be imported into other Python scripts to use its functions and variables.

---

## **2. Why Use Modules?**
✅ **Code Reusability** – Write once, use multiple times.  
✅ **Organized Code** – Avoid clutter in a single file.  
✅ **Encapsulation** – Hide implementation details and expose only necessary parts.  
✅ **Namespace Management** – Avoids name conflicts by organizing functions and classes separately.

---

## **3. Types of Modules**
Python has three types of modules:
1. **Built-in Modules** – Provided by Python (e.g., `math`, `os`, `sys`).
2. **User-defined Modules** – Created by the user.
3. **Third-party Modules** – Installed from external sources (`pip`).

---

## **4. Creating and Importing Modules**
### **A. Creating a User-Defined Module**
A module is simply a **Python file** (`.py`) that contains functions, variables, and classes.

#### **Example: Creating a Module (`mymodule.py`)**

# **Understanding `__name__ == "__main__"` in Python**
---
## **1. Introduction**
In Python, `__name__ == "__main__"` is a special construct that allows us to differentiate between **executing a script directly** and **importing it as a module** into another script. This is important for organizing code properly, avoiding unwanted executions, and enabling reusability.

---

## **2. What is `__name__` in Python?**
Every Python module (script) has a built-in variable called `__name__`. This variable represents the **name of the module** in which the code is running.

- When a Python file is run **directly**, `__name__` is set to `"__main__"`.
- When the same file is **imported as a module** into another script, `__name__` is set to the module’s actual name (filename without `.py`).

#### **Example: Checking `__name__` in a Python Script (`script.py`)**
```python
print(__name__)
```
✅ **If you run `script.py` directly**:
```bash
python script.py
```
Output:
```
__main__
```

✅ **If you import `script.py` into another script**:
```python
import script
```
Output:
```
script
```
This shows that `__name__` changes when the script is imported.

---

## **3. Why Use `if __name__ == "__main__"`?**
The `if __name__ == "__main__"` construct is used to **prevent certain code from running when the script is imported as a module**.

### **Example Without Using `if __name__ == "__main__"`**
```python
# module.py
def greet():
    print("Hello from module!")

print("This will run no matter what!")
```
✅ **Running directly (`python module.py`)**:
```
This will run no matter what!
```

✅ **Importing into another script**:
```python
import module
```
Output:
```
This will run no matter what!
```
⚠️ **Problem**: The print statement runs even when the file is imported.

---

### **Example Using `if __name__ == "__main__"`**
```python
# module.py
def greet():
    print("Hello from module!")

if __name__ == "__main__":
    print("This will run only when executed directly!")
```
✅ **Running directly (`python module.py`)**:
```
This will run only when executed directly!
```

✅ **Importing into another script (`import module`)**:
✅ **No output!** (Because the `if` condition prevents execution)

**🛠️ Key Benefits:**
1. **Prevents unwanted execution** when the script is imported.
2. **Separates reusable functions from main script execution.**

---

## **4. Real-World Use Cases**
### **A. Running Code Only When Script is Executed Directly**
```python
# script.py
def main():
    print("Main function executing...")

if __name__ == "__main__":
    main()
```
This ensures that `main()` only runs when `script.py` is executed directly.

---

### **B. Defining Reusable Functions Without Unnecessary Execution**
```python
# calculations.py
def add(a, b):
    return a + b

if __name__ == "__main__":
    print("Running test cases...")
    print(add(2, 3))  # Test case
```
✅ **Direct execution (`python calculations.py`)** runs the test case.  
✅ **Importing into another script (`import calculations`)** does not run test cases.

---

### **C. Using `__name__ == "__main__"` in Web Applications**
Many frameworks, such as **Flask**, use this construct.

```python
from flask import Flask

app = Flask(__name__)

@app.route('/')
def home():
    return "Hello, Flask!"

if __name__ == "__main__":
    app.run(debug=True)
```
✅ **Ensures Flask server runs only when script is executed directly.**

---

## **5. How `__name__` Works Internally**
### **Case 1: Running as a Script**
```python
# file: myscript.py
print(__name__)  # Output: "__main__"
```

### **Case 2: Importing as a Module**
```python
# file: main.py
import myscript  # Output: "myscript"
```
Internally, Python assigns the value of `__name__` based on whether it is **executed directly** or **imported**.

---

## **6. Common Mistakes to Avoid**
❌ **Forgetting `if __name__ == "__main__"`**
```python
def main():
    print("Executing...")

main()  # This runs even when imported!
```
✅ **Correct Approach**
```python
def main():
    print("Executing...")

if __name__ == "__main__":
    main()  # Runs only if executed directly
```

| `"__main__"` | Value of `__name__` when script is run directly |
| Module Name | Value of `__name__` when script is imported |
| `if __name__ == "__main__"` | Ensures code runs only when script is executed directly |
| Benefits | Prevents unwanted execution, allows code reusability |
| Use Cases | Scripts, test cases, Flask apps, command-line tools |

✅ **Using `if __name__ == "__main__"` helps keep Python programs modular, reusable, and well-structured!** 🚀

# **📦 Python Packages – A Complete Guide**  

## **🔹 What is a Package in Python?**  
A **package** in Python is a **collection of related modules** grouped in a directory. It helps in organizing **large-scale projects** by structuring them into smaller, manageable units.

### **🛠 Example:**
```
mypackage/      # This is the package (a directory)
│── __init__.py  # Required to make it a package
│── module1.py   # A Python module
│── module2.py   # Another Python module
```

In this structure:  
- `mypackage/` is a **package** (folder).  
- `module1.py` and `module2.py` are **modules** inside the package.  
- `__init__.py` makes the directory a **package**.  

---

## **🔹 Why Use Packages?**
- **Organizes Code** → Prevents clutter in large projects.  
- **Encapsulates Related Functionality** → Groups related modules together.  
- **Encourages Reusability** → Modules inside a package can be reused in multiple projects.  
- **Prevents Name Conflicts** → Similar module names in different packages don’t interfere with each other.  

---

## **🔹 Creating a Package in Python**
To create a package, follow these **3 steps**:

### **📌 Step 1: Create a Package Directory**
Let's create a package named `mypackage`.  
```
mypackage/
│── __init__.py   # Makes this directory a package
│── math_utils.py
│── string_utils.py
```

### **📌 Step 2: Write Code in Modules**
#### **math_utils.py**
```python
def add(a, b):
    return a + b

def multiply(a, b):
    return a * b
```

#### **string_utils.py**
```python
def to_uppercase(text):
    return text.upper()
```

### **📌 Step 3: Create `__init__.py`**
`__init__.py` is a special file that **makes a directory a package**.  
**(Since Python 3.3, it is optional, but good practice to include it.)**

#### **mypackage/__init__.py**
```python
from .math_utils import add, multiply
from .string_utils import to_uppercase
```
This allows importing functions directly from `mypackage`.

---

## **🔹 Importing from a Package**
### **Method 1: Import the Entire Package**
```python
import mypackage.math_utils

print(mypackage.math_utils.add(3, 5))  # Output: 8
```

### **Method 2: Import Specific Functions**
```python
from mypackage.math_utils import multiply

print(multiply(3, 5))  # Output: 15
```

### **Method 3: Using `__init__.py` for Direct Imports**
Since we defined `__init__.py` like this:
```python
from .math_utils import add, multiply
from .string_utils import to_uppercase
```
Now we can import directly from `mypackage`:
```python
from mypackage import add, to_uppercase

print(add(10, 5))  # Output: 15
print(to_uppercase("hello"))  # Output: HELLO
```

---

## **🔹 Nested Packages (Sub-Packages)**
A package can contain **sub-packages** (nested directories with `__init__.py`).

### **Example:**
```

│── __init__.py
│── utilities/        # Sub-package
│   │── __init__.py
│   │── file_utils.py
```

To import from a **sub-package**:
```python

```

---

## **🔹 Absolute Import vs Relative Import**
### **🔹 Absolute Import**
Imports using **full path** from the root.
```python
import mypackage.math_utils
from mypackage.string_utils import to_uppercase
```

### **🔹 Relative Import**
Imports using **dot notation** inside a package.
```python
from .math_utils import add
from .string_utils import to_uppercase
```
- **`.`** (single dot) → Refers to the current package  
- **`..`** (double dot) → Refers to the parent package

```python
from setuptools import setup, find_packages

setup(
    name="mypackage",  # Package name
    version="0.1.0",   # Version number
    author="Your Name",
    author_email="your.email@example.com",
    description="A simple math and utility package",
    long_description=open("README.md").read(),
    long_description_content_type="text/markdown",
    url="https://github.com/yourusername/mypackage",  # GitHub URL
    packages=find_packages(),  # Automatically find package directories
    classifiers=[  # Metadata (optional)
        "Programming Language :: Python :: 3",
        "License :: OSI Approved :: MIT License",
        "Operating System :: OS Independent",
    ],
    python_requires=">=3.6",  # Minimum Python version required
)
```

### **📌 Difference Between File, Module, Package, Library, and Framework in Python**  

| Concept     | Definition | Example |
|------------|-----------|---------|
| **File** | A single document stored on a system, containing code, text, or data. | `script.py`, `data.txt` |
| **Module** | A single Python file containing functions, classes, or variables that can be imported. | `math.py`, `os.py` |
| **Package** | A collection of multiple related Python modules, organized in directories with `__init__.py`. | `numpy`, `pandas` |
| **Library** | A collection of related packages and modules providing reusable functionality. | `requests`, `matplotlib` |
| **Framework** | A complete, structured environment that provides tools and rules for application development. | `Django`, `Flask` |

---

## **🔹 1. File in Python**
A **file** is the **smallest unit** in Python that stores code or data.  
It can be a Python script (`.py`), a text file (`.txt`), or any other type.

### **Example of a Python File (`script.py`)**
```python
print("Hello, World!")
```
This file can be run using:
```bash
python script.py
```

---

## **🔹 2. Module in Python**
A **module** is a **single Python file (`.py`)** that contains reusable functions, classes, or variables.  
Modules help **organize code** and **avoid repetition**.

### **Example of a Module (`math_operations.py`)**
```python
def add(a, b):
    return a + b

def subtract(a, b):
    return a - b
```
### **Using the Module in Another Python File**
```python
import math_operations

print(math_operations.add(5, 3))  # Output: 8
```
### **Types of Modules**
- **Built-in Modules** (pre-installed) → `math`, `os`, `sys`
- **User-defined Modules** → Custom files written by developers
- **Third-party Modules** → Installed via `pip` (`requests`, `numpy`)

---

## **🔹 3. Package in Python**
A **package** is a **collection of modules**, grouped together in a directory.  
Every package contains a special `__init__.py` file, which **marks the directory as a package**.

### **Example of a Package (`mypackage/`)**
```
mypackage/
│── __init__.py      # Marks this as a package
│── operations.py    # Module inside package
│── utils.py         # Another module inside package
```

### **Using the Package in Another Python File**
```python
from mypackage.operations import add

print(add(10, 5))  # Output: 15
```

### **Difference Between Module & Package**
| **Module** | **Package** |
|------------|------------|
| A **single** `.py` file | A **collection** of modules |
| Cannot contain sub-modules | Can contain multiple modules |
| Example: `math.py` | Example: `numpy` (a package with many modules) |

---

## **🔹 4. Library in Python**
A **library** is a **collection of multiple packages and modules** that provide a specific functionality.  
Python libraries contain **pre-written code** to speed up development.

### **Examples of Popular Python Libraries**
| Library | Purpose |
|---------|---------|
| `NumPy` | Scientific computing, arrays |
| `Pandas` | Data analysis |
| `Matplotlib` | Data visualization |
| `Requests` | HTTP requests |
| `TensorFlow` | Machine Learning |

### **Using a Library**
```python
import requests

response = requests.get("https://example.com")
print(response.status_code)
```
A **library is bigger than a package** because it includes **multiple related packages**.

---

## **🔹 5. Framework in Python**
A **framework** is a **structured environment** for application development.  
Unlike a library, a framework **dictates the architecture** of the project.

### **Key Differences Between Library & Framework**
| **Library** | **Framework** |
|------------|--------------|
| Provides **specific functionality** | Provides a **complete structure** for development |
| Developers decide when & how to use it | The framework **controls the flow** of the program |
| Example: `requests` (used anywhere) | Example: `Django` (follows MVC structure) |

### **Popular Python Frameworks**
| Framework | Use Case |
|-----------|---------|
| `Django` | Full-stack web development |
| `Flask` | Lightweight web development |
| `FastAPI` | API development |
| `PyTorch` | Deep Learning |
| `Selenium` | Web automation |

### **Library vs Framework in Python**

Both libraries and frameworks are tools designed to make your development process easier by providing reusable code, but they work in different ways. Here's a detailed comparison:

#### **Basic Definition:**
- **Library**: A collection of pre-written code, functions, and classes that you can use to perform common tasks. You **call** the library functions as needed.
- **Framework**: A larger, more comprehensive set of code that dictates the structure and flow of the application. In a framework, you **follow** the framework's rules and the framework calls your code when necessary.

---

#### **Key Differences:**

1. **Control Flow**:
   - **Library**: You, the programmer, are in control. You decide when and how to call the library. The flow of the program is managed by you.
   - **Framework**: The framework controls the flow of the program. It calls your code at certain points, and you fill in the specific details at those points (often through callbacks or specific methods).

2. **Code Structure**:
   - **Library**: You define the structure of the code, as you decide how and when to integrate the library into your project.
   - **Framework**: The framework enforces the structure of the code. It typically comes with predefined rules and conventions (like folder structures, naming conventions, etc.) that you have to follow.

3. **Inversion of Control** (The "Hollywood Principle"):
   - **Library**: "You call us when you need us."
   - **Framework**: "Don't call us, we'll call you." The framework dictates the flow and calls your code at specific points.

---

#### **Real-World Analogies**:

- **Library (e.g., Python's NumPy or requests)**:
  - Imagine you’re cooking in your kitchen. A **library** is like a shelf full of spices and tools (like a blender or a knife). You decide when and how to use them in your recipe. You are still in control of the cooking process.
  
- **Framework (e.g., Django or Flask)**:
  - A **framework** is like following a cookbook recipe. It provides you with a step-by-step guide to cooking (or building an application). It defines the structure, ingredients, and the overall process. You only focus on following the steps provided by the framework, and it will call on you to add specific ingredients (code) when needed.

---

#### **Examples in Python**:

- **Library Examples**:
  - **NumPy**: A library that provides functions for numerical computations. You control when to call its functions, and you define the structure of your program.
  - **requests**: A library for making HTTP requests. You call its functions when needed to interact with APIs or web services.

- **Framework Examples**:
  - **Django**: A web framework that helps you build web applications by providing the structure for things like routing, authentication, database handling, and more. You follow its structure and plug in your business logic where needed.
  - **Flask**: Another web framework (lighter than Django) that helps you build web apps. It gives you more freedom, but still imposes structure in how routes and requests are handled.

---

#### **Summary**:

- **Library**: You have the control. You call specific functions when you need them.
- **Framework**: The framework controls the flow of the application. You fill in the required pieces at predefined points.

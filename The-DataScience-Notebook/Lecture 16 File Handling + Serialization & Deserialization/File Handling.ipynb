# **File Handling in Python**



File handling in Python allows you to create, open, read, write, append, and manage files on your computer. Python provides built-in functions and libraries to interact with files efficiently.

---

## **Why File Handling?**
1. **Persistence**: Store data beyond the runtime of the program.
2. **Data Sharing**: Read from or write to files for data exchange.
3. **Logging**: Keep logs of application activities.
4. **Backup**: Store backups of processed data.

---

## **Types of Files in Python**
1. **Text Files**:
   - Contain plain text.
   - Encoding like UTF-8 is common.
   - Examples: `.txt`, `.csv`, `.json`.

2. **Binary Files**:
   - Contain data in binary format (0s and 1s).
   - Examples: Images, videos, `.exe`, `.bin`.

---

1. **Open a File**
2. **Read Data**

4. **Append Data**
5. **Close a File**

## **Opening a File**
Python provides the `open()` function to open files.

### **Syntax**:
```python
file = open(file_name, mode, encoding)
```

| **Mode** | **Description**                                                                 |
|----------|---------------------------------------------------------------------------------|
| `r`      | Read mode (default). Opens the file for reading.                                |
| `w`      | Write mode. Creates a new file or overwrites an existing file.                  |
| `a`      | Append mode. Adds content to the end of the file.                               |
| `x`      | Exclusive creation mode. Fails if the file already exists.                      |
| `rb`     | Read mode for binary files.                                                    |
| `wb`     | Write mode for binary files.                                                   |
| `ab`     | Append mode for binary files.                                                  |

---

## **Reading a File**
To read the content of a file, you can use:

### **Methods**:
1. **`read()`**: Reads the entire file.
2. **`readline()`**: Reads a single line at a time.
3. **`readlines()`**: Reads all lines into a list.

### **Example**:
```python
file = open("example.txt", "r")
print(file.read())       # Reads the entire file
file.seek(0)             # Move the cursor to the beginning
print(file.readline())   # Reads the first line
print(file.readlines())  # Reads all lines into a list
file.close()
```

# **Complete Notes on `open()` in Python**



The `open()` function is the primary method in Python for working with files. It is used to open a file for various operations such as reading, writing, appending, and more.

---

## **Syntax**

```python
open(file, mode='r', buffering=-1, encoding=None, errors=None, newline=None, closefd=True, opener=None)
```

### **Parameters of `open()`**

| **Parameter** | **Description**                                                                                     |
|---------------|-----------------------------------------------------------------------------------------------------|
| `file`        | The name or path of the file to open. Can be a relative or absolute path.                           |
| `mode`        | The mode in which the file is opened (e.g., `r`, `w`, `a`). Default is `'r'` (read).                |
| `buffering`   | Controls buffering of the file (integer). Default is `-1` (uses system defaults).                   |
| `encoding`    | Encoding format (e.g., `'utf-8'`). Used for text files.                                             |
| `errors`      | Specifies how encoding/decoding errors are handled (e.g., `'strict'`, `'ignore'`, `'replace'`).     |
| `newline`     | Controls how newline characters are handled in text mode.                                           |
| `closefd`     | Only applicable for file descriptors. If `False`, the file descriptor is not closed upon file close.|
| `opener`      | A custom callable that opens the file.                                                             |

---

## **File Modes in `open()`**

The `mode` parameter specifies the purpose of opening the file. 

| **Mode** | **Description**                                                                                     |
|----------|-----------------------------------------------------------------------------------------------------|
| `'r'`    | Read mode. Opens the file for reading (default).                                                    |
| `'w'`    | Write mode. Creates a new file or overwrites the existing file.                                     |
| `'a'`    | Append mode. Adds content to the end of the file without overwriting.                               |
| `'x'`    | Exclusive creation mode. Fails if the file already exists.                                          |
| `'b'`    | Binary mode. Used for binary files. Can be combined with other modes.                               |
| `'t'`    | Text mode. Default mode for text files. Can be combined with other modes.                           |
| `'+'`    | Update mode. Allows both reading and writing.                                                       |

### **Mode Combinations**

| **Mode**   | **Meaning**                                      |
|------------|--------------------------------------------------|
| `'rb'`     | Read binary data.                               |
| `'wb'`     | Write binary data.                              |
| `'ab'`     | Append binary data.                             |
| `'r+'`     | Read and write.                                 |
| `'w+'`     | Write and read (overwrites file).               |
| `'a+'`     | Append and read.                                |

---

## **Return Value of `open()`**

- The `open()` function returns a **file object**.
- Depending on the mode, the file object allows various operations like reading, writing, and appending.

---

## **Buffering in `open()`**

The `buffering` parameter controls the buffering policy:
1. **Default (`-1`)**: Uses system defaults.
2. **`0`**: No buffering (only in binary mode).
3. **`1`**: Line buffering (used in text mode).
4. **`>1`**: Fixed-size buffering (size is specified by the integer).

---

## **Encoding in `open()`**

- Text files use encodings like UTF-8, ASCII, etc.
- Default encoding varies by system.
- To ensure compatibility, explicitly specify `encoding='utf-8'`.

# **Comprehensive Notes on `writelines()` in Python**



The `writelines()` method is a file object method in Python used to write multiple lines of text or binary data to a file. This method is particularly useful when you have a sequence of strings (such as a list) and want to write them all at once to a file.

---

## **Syntax**

```python
file.writelines(lines)
```

### **Parameters**
- **`lines`**: A sequence (list, tuple, or any iterable) containing strings or bytes. Each element in the sequence represents a line to be written to the file.

### **Return Value**
- `None`. The method does not return anything.

---

## **Key Features**
1. **Efficient for Batch Writing**: `writelines()` is efficient for writing multiple lines at once instead of calling `write()` repeatedly.
2. **No Newline Character by Default**: Unlike `print()`, `writelines()` does not automatically add newline characters (`\n`) between lines. You need to include them explicitly in the strings.
3. **Supports Text and Binary Files**: Can be used for both text (`'t'` mode) and binary (`'b'` mode) files.
4. **Iterable Requirement**: The input must be an iterable, such as a list, tuple, or generator.

---

## **Usage Examples**

### **Writing Multiple Lines to a File**
```python
lines = ["Line 1\n", "Line 2\n", "Line 3\n"]

# Open file in write mode
with open("example.txt", "w") as file:
    file.writelines(lines)

# Output in 'example.txt':
# Line 1
# Line 2
# Line 3
```

---

### **Writing Without Explicit Newline Characters**
If newline characters are not included in the strings, the lines will be written consecutively without breaks.

```python
lines = ["Line 1", "Line 2", "Line 3"]

with open("example.txt", "w") as file:
    file.writelines(lines)

# Output in 'example.txt':
# Line 1Line 2Line 3
```

---

### **Using a Generator**
You can use a generator to dynamically write lines to the file.

```python
def generate_lines():
    for i in range(1, 4):
        yield f"Generated Line {i}\n"

with open("example.txt", "w") as file:
    file.writelines(generate_lines())

# Output in 'example.txt':
# Generated Line 1
# Generated Line 2
# Generated Line 3
```

---

### **Writing Binary Data**
For binary files, the input iterable must contain `bytes` objects.

```python
lines = [b"Binary Line 1\n", b"Binary Line 2\n"]

with open("example.bin", "wb") as file:
    file.writelines(lines)

# The file 'example.bin' will contain:
# Binary Line 1
# Binary Line 2
```

---

## **Handling Newline Characters**

### **Manually Adding Newlines**
If your lines do not already include newline characters (`\n`), you can modify them before using `writelines()`.

```python
lines = ["Line 1", "Line 2", "Line 3"]
lines_with_newlines = [line + "\n" for line in lines]

with open("example.txt", "w") as file:
    file.writelines(lines_with_newlines)

# Output in 'example.txt':
# Line 1
# Line 2
# Line 3
```

### **Using `map()` to Add Newlines**
An alternative way to append newlines is by using the `map()` function.

```python
lines = ["Line 1", "Line 2", "Line 3"]

with open("example.txt", "w") as file:
    file.writelines(map(lambda line: line + "\n", lines))

# Output in 'example.txt':
# Line 1
# Line 2
# Line 3
```

---

## **Comparison: `write()` vs `writelines()`**

| Feature                 | `write()`                                    | `writelines()`                                  |
|-------------------------|----------------------------------------------|------------------------------------------------|
| **Input**               | Single string or bytes                      | Iterable of strings or bytes                   |
| **Newline Handling**    | Manual (no automatic newline)               | Manual (no automatic newline)                 |
| **Use Case**            | Writing a single line or block of text      | Writing multiple lines or strings at once     |
| **Performance**         | Slower when writing multiple lines          | Faster for writing multiple lines             |

### **Example: Using `write()` and `writelines()`**
```python
# Using write()
with open("example.txt", "w") as file:
    for line in ["Line 1\n", "Line 2\n", "Line 3\n"]:
        file.write(line)

# Using writelines()
with open("example.txt", "w") as file:
    file.writelines(["Line 1\n", "Line 2\n", "Line 3\n"])
```

---

## **Common Mistakes**

1. **Missing Newline Characters**: Forgetting to include `\n` leads to concatenated lines.
   ```python
   lines = ["Line 1", "Line 2"]
   with open("example.txt", "w") as file:
       file.writelines(lines)
   # Output: Line 1Line 2
   ```

2. **Using Non-Iterable Input**: Passing a single string or non-iterable object results in a `TypeError`.
   ```python
   with open("example.txt", "w") as file:
       file.writelines("This is a single string")  # TypeError
   ```

3. **Inconsistent Data Types**: Mixing `str` and `bytes` in the same iterable raises a `TypeError`.
   ```python
   lines = ["Line 1\n", b"Binary Line 2\n"]
   with open("example.txt", "w") as file:
       file.writelines(lines)  # TypeError
   ```

---

## **Advanced Tips**

1. **Using `join()` for Simplicity**: If you want to add a single newline character between lines, you can use `join()` with `write()` instead of `writelines()`.
   ```python
   lines = ["Line 1", "Line 2", "Line 3"]
   with open("example.txt", "w") as file:
       file.write("\n".join(lines))
   ```

2. **Handling Large Data**: For large datasets, use generators with `writelines()` to avoid loading all data into memory at once.

3. **Using Context Managers**: Always use `with` statements to handle files to avoid resource leaks.

---

## **Performance Considerations**
- **`writelines()` vs `write()`**: For writing multiple lines, `writelines()` is faster since it reduces the number of I/O operations.
- **Buffering**: Use buffering appropriately (`buffering=1024` for example) for performance when working with large files.

---

## **Key Points to Remember**
1. `writelines()` does not add newline characters; include them explicitly in the strings.
2. Input must be an iterable containing only `str` (text mode) or `bytes` (binary mode) objects.
3. Use `map()` or list comprehensions to preprocess lines if needed.
4. For single-line writing, prefer `write()`, but for multiple lines, `writelines()` is more efficient.

# **Comprehensive Notes on Reading from Files in Python**



Reading from files is a crucial aspect of file handling in Python, allowing you to access and process data stored in text or binary files. This guide will cover all aspects of reading files in Python, including the methods, modes, and best practices.

---

## **Opening a File for Reading**

To read from a file, you must open it in one of the following modes:
- **`'r'`**: Opens the file in text mode for reading (default).
- **`'rb'`**: Opens the file in binary mode for reading.

### **Syntax**
```python
file_object = open("filename", mode)
```

Always ensure you close the file after reading to free system resources. Use `with` statements for automatic file closing.

---

## **Methods for Reading Files**

### 1. **`read()`**
Reads the entire content of the file as a single string or bytes object.

#### **Syntax**
```python
file_object.read(size)
```

#### **Parameters**
- **`size`** (optional): Number of characters (or bytes) to read. If omitted, reads the entire file.

#### **Example**
```python
with open("example.txt", "r") as file:
    content = file.read()
    print(content)
```

#### **Features**
- If `size` is specified, only that many characters/bytes are read.
- If `size` is not specified, it reads until EOF (End of File).

#### **Binary Example**
```python
with open("example.bin", "rb") as file:
    binary_data = file.read(10)  # Reads 10 bytes
    print(binary_data)
```

---

### 2. **`readline()`**
Reads one line at a time.

#### **Syntax**
```python
file_object.readline(size=-1)
```

#### **Parameters**
- **`size`** (optional): Maximum number of characters to read. Default is `-1` (reads the entire line).

#### **Example**
```python
with open("example.txt", "r") as file:
    first_line = file.readline()
    print(first_line)
```

#### **Behavior**
- Reads until a newline character (`\n`) or EOF is reached.
- Returns an empty string if the end of the file is reached.

---

### 3. **`readlines()`**
Reads all lines of the file into a list, with each line as a separate string.

#### **Syntax**
```python
file_object.readlines(hint=-1)
```

#### **Parameters**
- **`hint`** (optional): Total number of bytes to read. Default is `-1` (reads all lines).

#### **Example**
```python
with open("example.txt", "r") as file:
    lines = file.readlines()
    for line in lines:
        print(line.strip())
```

---

### 4. **Iterating Over the File Object**
A file object is iterable, allowing you to loop over its lines directly.

#### **Example**
```python
with open("example.txt", "r") as file:
    for line in file:
        print(line.strip())
```

---

## **File Modes for Reading**

| Mode   | Description                                                                 |
|--------|-----------------------------------------------------------------------------|
| `'r'`  | Read text file. Raises `FileNotFoundError` if the file does not exist.      |
| `'rb'` | Read binary file. Raises `FileNotFoundError` if the file does not exist.    |
| `'r+'` | Read and write. Does not create the file if it does not exist.              |

---

When reading large files, avoid loading the entire file into memory. Use the following techniques:

### **Reading in Chunks**


### **Line-by-Line Reading**
```python

```

## **File Positioning**

The `tell()` and `seek()` methods help manage file reading positions.

### **`tell()`**
Returns the current position of the file pointer (in bytes).

```python
with open("example.txt", "r") as file:
    print(file.tell())  # Outputs the current position
```

### **`seek()`**
Moves the file pointer to a specific position.

#### **Syntax**
```python
file_object.seek(offset, whence)
```

- **`offset`**: Number of bytes to move.
- **`whence`**: Reference point:
  - `0`: Start of the file (default).
  - `1`: Current position.
  - `2`: End of the file.

#### **Example**
```python

    file.seek(5)  # Moves to the 5th byte
    print(file.read())  # Reads from the 5th byte
```

# Problems with Working in Text Mode in Python



Text mode (`'r'`, `'w'`, `'a'`, etc.) is commonly used to handle files containing human-readable data (e.g., `.txt`, `.csv`, `.json`). While text mode simplifies operations with encoded character data, it comes with limitations and potential pitfalls that can cause issues in certain scenarios.

---

## **Key Problems with Text Mode**

### 1. **Encoding Issues**
- Text mode relies on the default or specified encoding to interpret characters.
- **Problem**: If the file is saved with a different encoding (e.g., UTF-16 instead of UTF-8), decoding errors may occur.
- **Error**: 
  - `UnicodeDecodeError` when reading files.
  - `UnicodeEncodeError` when writing files.
  
#### **Example**
```python
with open("file.txt", "r", encoding="utf-8") as file:
    content = file.read()  # May raise UnicodeDecodeError if the file isn't UTF-8.
```

---

### 2. **Line Ending Variations**
- Different operating systems use different line-ending conventions:
  - **Windows**: `\r\n`
  - **Unix/Linux/Mac**: `\n`
  - **Old Mac (pre-OSX)**: `\r`
- **Problem**: Text mode automatically normalizes line endings to `\n`. This can cause compatibility issues when files are shared between systems or when binary file manipulation is needed.

#### **Example**
```python
# Line endings may vary if opened in text mode on different platforms
with open("file.txt", "r") as file:
    for line in file:
        print(repr(line))  # Line endings may appear inconsistent
```

---

### 3. **Performance Bottlenecks**
- Text mode performs implicit character encoding/decoding and newline normalization.
- **Problem**: Slower than binary mode when working with large files or raw data streams.

#### **Example**
```python
with open("large_file.txt", "r") as file:
    content = file.read()  # Decoding each character can take longer for massive files
```

---

### 4. **Corruption of Binary Files**
- Text mode interprets the file content as encoded characters, which is unsuitable for binary data (e.g., images, audio, or executables).
- **Problem**: Opening binary files in text mode may corrupt the data by misinterpreting non-text bytes.

#### **Example**
```python
# Trying to open an image in text mode
with open("image.jpg", "r") as file:
    content = file.read()  # Corrupts binary data, leading to garbage output
```

---

### 5. **Truncation of Null Bytes (`\x00`)**
- In text mode, null bytes (`\x00`) might not be handled correctly because they are typically non-printable in text data.
- **Problem**: Truncation or misinterpretation of data containing null bytes.

#### **Example**
```python
# Null byte in a text file
with open("file.txt", "r") as file:
    content = file.read()
    print(content)  # May behave unexpectedly with null bytes
```

---

### 6. **Locale Dependency**
- Text mode uses the system's default locale for encoding if no encoding is specified.
- **Problem**: The behavior of text operations may change across different systems or environments with different locale settings.

#### **Example**
```python
# Locale-dependent encoding
with open("file.txt", "r") as file:
    content = file.read()  # May vary if the locale uses a different default encoding
```

---

### 7. **File Size Misinterpretation**
- File sizes in text mode can be misleading due to encoding/decoding.
- **Problem**: The number of characters read does not necessarily match the file's byte size on disk.

#### **Example**
```python
with open("file.txt", "r") as file:
    content = file.read()
    print(len(content))  # Length in characters, not bytes
```

---

### 8. **Limited Control Over Data**
- Text mode lacks precise control over file content, such as handling raw bytes or exact file offsets.
- **Problem**: Difficult to work with structured data formats (e.g., fixed-width binary formats or custom protocols).

#### **Example**
```python
# Cannot precisely handle raw data in text mode
with open("data.bin", "r") as file:
    file.seek(10)  # Offset behavior may be unpredictable due to encoding
```

---

### 9. **Inconsistencies in Data Representation**
- Different systems and Python versions might handle characters like newline or whitespace inconsistently in text mode.
- **Problem**: Misalignment of data expectations, especially when processing text programmatically.

---

## **When to Avoid Text Mode**
- **Binary Data**: For images, audio, video, executables, or any file containing non-textual data.
- **Cross-Platform Compatibility**: When line-ending or encoding consistency is critical.
- **High Performance**: When working with large files or raw data streams.

---

## **Best Practices**
1. **Use Binary Mode (`'rb'`, `'wb'`)**: When working with non-textual data or when precise control is required.
2. **Specify Encoding**: Always specify the encoding explicitly (e.g., `encoding="utf-8"`) to avoid surprises.
3. **Handle Exceptions**: Use try-except blocks to catch and handle encoding/decoding errors.
4. **Normalize Line Endings**: Use libraries like `os` or `io` to standardize line endings when necessary.
5. **Test Across Platforms**: Ensure your code behaves as expected on all target platforms.

---

By understanding these limitations and adopting the best practices, you can minimize problems while working in text mode in Python.

## 1. What is Serialization and Deserialization?

### 1.1 Definition

- **Serialization**: The process of converting a Python object into a byte stream or a string representation that can be stored or transmitted.
- **Deserialization**: The reverse process of converting the byte stream or string back into a Python object.

### 1.2 Why Use Serialization?

- **Persistence**: Save objects to a file or database for later retrieval.
- **Data Exchange**: Send objects over a network between different programs or systems.
- **Caching**: Store preprocessed data for faster retrieval.
- **Logging and Debugging**: Record the state of an object at a specific point in time.

---

## 2. Python's Built-in Serialization Modules

Python provides several built-in modules for serialization:

### 2.1 `pickle` Module

- **Purpose**: Serialize and deserialize Python objects to and from binary format.
- **Supports**: Nearly all Python data types, including custom classes.
- **File Formats**: Binary data, not human-readable.
- **Usage**:

  ```python
  import pickle

  data = {'key': 'value'}
  serialized_data = pickle.dumps(data)
  deserialized_data = pickle.loads(serialized_data)
  ```

### 2.2 `json` Module

- **Purpose**: Serialize and deserialize Python objects to and from JSON format.
- **Supports**: Basic data types (numbers, strings, lists, dictionaries).
- **File Formats**: Text-based, human-readable.
- **Usage**:

  ```python
  import json

  data = {'key': 'value'}
  serialized_data = json.dumps(data)
  deserialized_data = json.loads(serialized_data)
  ```

### 2.3 `marshal` Module

- **Purpose**: Internal serialization for Python bytecode.
- **Supports**: Basic data types.
- **Note**: Not recommended for general use as the format is not guaranteed to be stable across Python versions.

### 2.4 `shelve` Module

- **Purpose**: Provides a simple persistent storage option with a dictionary-like API.
- **Usage**:

  ```python
  import shelve

  with shelve.open('mydata.db') as db:
      db['key'] = {'value': 42}
      print(db['key'])
  ```

## 3. Third-Party Serialization Libraries

In addition to built-in modules, Python has rich support for third-party serialization libraries.

### 3.1 YAML (`PyYAML`)

- **Purpose**: Serialize and deserialize Python objects to and from YAML format.
- **Advantages**: Human-readable, supports complex data structures.

  ```python
  import yaml

  data = {'key': 'value'}
  serialized_data = yaml.dump(data)
  ```

### 3.2 MessagePack (`msgpack`)

- **Purpose**: Binary serialization format that's more efficient than JSON.
- **Advantages**: Compact, fast, supports streaming.

  ```python
  import msgpack

  deserialized_data = msgpack.unpackb(serialized_data)
  ```

- **Purpose**: Language-neutral, platform-neutral extensible mechanism for serializing structured data.
- **Advantages**: Efficient, cross-language compatibility.
- **Usage**:

  Define a `.proto` file and use the `protoc` compiler to generate Python code.

### 3.4 Apache Avro (`avro`)

- **Purpose**: Data serialization system with rich data structures.
- **Advantages**: Schema evolution, compact binary format.
- **Usage**:

  ```python
  import avro.schema
  ```

#### 4.2.3 Handling Non-Serializable Objects

- **Problem**: The `json` module cannot serialize custom objects by default.
- **Solution**: Implement custom serialization.

```python
class MyClass:
    def __init__(self, value):
        self.value = value

def serialize(obj):
    if isinstance(obj, MyClass):
        return {'__MyClass__': True, 'value': obj.value}
    raise TypeError(f"Type {type(obj)} not serializable")

def deserialize(dct):
    if '__MyClass__' in dct:
        return MyClass(dct['value'])
    return dct

# Serialize
json_string = json.dumps(obj, default=serialize)

# Deserialize
loaded_obj = json.loads(json_string, object_hook=deserialize)
```

### 4.3 Custom Serialization

#### 4.3.1 Implementing `__getstate__` and `__setstate__`

- **Purpose**: Control how an object is pickled and unpickled.

```python
class MyClass:
    def __init__(self, value):
        self.value = value
        self._transient = 'temporary data'

    def __getstate__(self):
        state = self.__dict__.copy()
        del state['_transient']
        return state

    def __setstate__(self, state):
        self.__dict__.update(state)
        self._transient = 'restored data'

# Serialize
with open('object.pkl', 'wb') as f:
    pickle.dump(obj, f)

# Deserialize
with open('object.pkl', 'rb') as f:
    loaded_obj = pickle.load(f)
print(loaded_obj._transient)
```

#### 4.3.2 Using `__reduce__` and `__reduce_ex__`

- **Purpose**: Provide complete control over the pickling process.

```python
class MyClass:
```

## 5. Security Considerations

### 5.1 Risks of `pickle`

- **Arbitrary Code Execution**: Unpickling data from untrusted sources can execute arbitrary code, leading to security vulnerabilities.
- **Example of Vulnerability**:

  ```python
  import pickle

  malicious_data = b"cos\nsystem\n(S'echo malicious code'\ntR."
  pickle.loads(malicious_data)  # Executes 'echo malicious code'
  ```

### 5.2 Safe Serialization Practices

- **Do Not Unpickle Untrusted Data**: Only unpickle data you trust.
- **Use Safe Modules**: For data exchange with untrusted sources, use `json` or other safe serialization formats.
- **Validate Data**: Implement strict validation when deserializing data.

## 6. Best Practices

### 6.1 Choosing the Right Serialization Format

- **Use `pickle`**:

  - When serializing Python-specific data.
  - For applications where both serialization and deserialization are under your control.
  - Not suitable for long-term storage (due to compatibility issues).

- **Use `json`**:

  - For interoperability with other systems/languages.
  - When human-readable format is important.
  - For simple data structures.

- **Use `yaml`**:

  - When you need a human-readable format that supports more complex data structures.

- **Use `protobuf`, `avro`, or `msgpack`**:

  - For efficient, cross-language data exchange.
  - When performance and compatibility are critical.

### 6.2 Optimizing Serialization Performance

- **Use Latest Protocols**: For `pickle`, use `protocol=pickle.HIGHEST_PROTOCOL`.
- **Avoid Redundant Data**: Only serialize necessary data.
- **Use Binary Formats**: For performance-critical applications, prefer binary formats over text-based formats.

The reason your set and tuple get stored and deserialized as a list is because JSON doesn't natively support the `set` or `tuple` data types. JSON's supported data types are:

- `str`
- `int`
- `float`
- `bool`
- `None`
- `list`
- `dict`

When you try to serialize a set or tuple using `json.dump()`, they get automatically converted to a list, which is the closest equivalent in JSON. When deserializing with `json.load()`, the resulting object will be a list as well.

To preserve the type (set or tuple) when serializing and deserializing, you can handle the conversion manually by converting the set or tuple to a JSON-serializable format and then converting it back upon loading.

Hereâ€™s an example of how you can do that:

### To serialize and deserialize a set or tuple correctly:
```python
import json

# Define a set (or tuple)
s = {1, 2, 3, 4, 5}

# Custom encoder to handle set serialization
def set_encoder(obj):
    if isinstance(obj, set):
        return list(obj)  # Convert set to list for serialization
    raise TypeError(f"Type {type(obj)} not serializable")

# Serialize the set to JSON
with open("demo.json", "w") as f:
    json.dump(s, f, default=set_encoder)

# Deserialize the list back to a set
def set_decoder(dct):
    if isinstance(dct, list):
        return set(dct)  # Convert list back to set
    return dct

# Deserialize the JSON back into a set
with open("demo.json", "r") as f:
    loaded_s = json.load(f, object_hook=set_decoder)
    print(loaded_s)
    print(type(loaded_s))
```

### Output:
```
{1, 2, 3, 4, 5}
<class 'set'>
```

This way, the set is serialized as a list but is deserialized back to a set. You can adapt this for tuples or other non-JSON-native types as well.

# Complete Notes on Serializing and Deserializing Custom Objects in Python



Serialization and deserialization of custom objects in Python involve converting objects of user-defined classes into a format that can be stored or transmitted (serialization) and converting them back into objects (deserialization). The process allows data persistence, inter-process communication, or sharing of data between different systems.

---

### **What is Serialization and Deserialization?**
1. **Serialization**: Converting an object into a storable format (e.g., JSON, bytes, or string).
   - Example: Saving an object to a file.
2. **Deserialization**: Reconstructing the object from the serialized data.
   - Example: Loading the object back into memory.

---

### **Challenges with Custom Objects**
- Built-in serialization libraries like `pickle` and `json` handle basic data types (integers, strings, lists, etc.) but do not automatically understand custom objects.
- Custom objects need additional handling to serialize their attributes properly.

---

### **Methods to Serialize Custom Objects**

#### **1. Using the `pickle` Module**
The `pickle` module is designed to serialize and deserialize Python objects, including custom objects.

**Example:**
```python
import pickle

class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

# Create an instance of Person
person = Person("Alice", 30)

# Serialize the object
with open("person.pkl", "wb") as file:
    pickle.dump(person, file)

# Deserialize the object
with open("person.pkl", "rb") as file:
    loaded_person = pickle.load(file)

print(loaded_person.name)  # Alice
print(loaded_person.age)   # 30
```

**Advantages of `pickle`**:
- Supports almost all Python objects, including custom ones.
- Easy to use.

**Disadvantages of `pickle`**:
- The format is Python-specific and not human-readable.
- Serialized data is not portable across programming languages.
- Potential security risks when loading untrusted data.

---

#### **2. Using the `json` Module**
The `json` module is widely used for human-readable and cross-platform serialization. However, it doesn't natively handle custom objects. To serialize custom objects, you need to provide custom encoding and decoding logic.

**Custom Encoder and Decoder Example:**
```python
import json

class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def __repr__(self):
        return f"Person(name={self.name}, age={self.age})"

# Custom JSON encoder
class CustomEncoder(json.JSONEncoder):
    def default(self, obj):
        if isinstance(obj, Person):
            return {'name': obj.name, 'age': obj.age}
        return super().default(obj)

# Custom JSON decoder
def custom_decoder(d):
    if 'name' in d and 'age' in d:
        return Person(d['name'], d['age'])
    return d

# Create an object
person = Person("Alice", 30)

# Serialize the object
with open("person.json", "w") as file:
    json.dump(person, file, cls=CustomEncoder)

# Deserialize the object
with open("person.json", "r") as file:
    loaded_person = json.load(file, object_hook=custom_decoder)

print(loaded_person)  # Person(name=Alice, age=30)
```

**Advantages of `json`**:
- Cross-platform compatibility.
- Human-readable format.

**Disadvantages of `json`**:
- Requires custom logic for custom objects.
- Cannot serialize all Python objects (e.g., functions or complex types).

---

#### **3. Using the `__dict__` Attribute**
In Python, objects have a `__dict__` attribute that stores their attributes as a dictionary. This makes it easy to serialize and deserialize objects manually.

**Example:**
```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def __repr__(self):
        return f"Person(name={self.name}, age={self.age})"

# Create an object
person = Person("Alice", 30)

# Serialize using __dict__
data = person.__dict__
with open("person.json", "w") as file:
    json.dump(data, file)

# Deserialize
with open("person.json", "r") as file:
    data = json.load(file)
    loaded_person = Person(**data)

print(loaded_person)  # Person(name=Alice, age=30)
```

**Advantages**:
- Simple and easy to implement.
- Works well for objects with straightforward attributes.

**Disadvantages**:
- Not suitable for objects with non-trivial attributes or methods.

---

### **Comparison of Methods**

| Method       | Pros                                | Cons                                | Use Cases                                     |
|--------------|-------------------------------------|-------------------------------------|----------------------------------------------|
| `pickle`     | Supports complex objects; easy to use | Python-specific; not human-readable | For Python-only applications needing complex serialization. |
| `json`       | Cross-platform; human-readable       | Needs custom encoding/decoding      | When compatibility with other systems is required. |
| `__dict__`   | Simple and intuitive                 | Limited to simple objects           | For quick serialization of simple objects.   |

---

### **Best Practices**
1. **Choose the right method for your use case**:
   - Use `pickle` for Python-only applications.
   - Use `json` for human-readable or cross-language compatibility.
2. **Ensure security when deserializing**:
   - Avoid untrusted data, especially with `pickle`, as it can execute arbitrary code.
3. **Use `__dict__` or custom encoders/decoders** for better control.

---

### **Common Pitfalls**
- **Mutable and Immutable Attributes**: Be cautious when serializing objects with mutable attributes, as changes after deserialization may affect other references.
- **Nested Objects**: Ensure custom encoding/decoding logic handles nested objects properly.
- **Inheritance**: If working with inherited classes, ensure the parent attributes are serialized/deserialized correctly.

---

### **Real-World Use Case**
Consider a scenario where you need to save the state of a game. You have multiple custom classes like `Player`, `Enemy`, and `Game`. Serialization allows you to save the state of these objects to a file, and deserialization enables you to reload the game later. 

**Example**:
```python
import json

class Player:
    def __init__(self, name, score):
        self.name = name
        self.score = score

class Game:
    def __init__(self, level, player):
        self.level = level
        self.player = player

class CustomEncoder(json.JSONEncoder):
    def default(self, obj):
        if isinstance(obj, Player):
            return {'name': obj.name, 'score': obj.score}
        elif isinstance(obj, Game):
            return {'level': obj.level, 'player': obj.player}
        return super().default(obj)

def custom_decoder(d):
    if 'name' in d and 'score' in d:
        return Player(d['name'], d['score'])
    if 'level' in d and 'player' in d:
        return Game(d['level'], d['player'])
    return d

player = Player("Alice", 100)
game = Game(3, player)

# Serialize
with open("game.json", "w") as file:
    json.dump(game, file, cls=CustomEncoder)

# Deserialize
with open("game.json", "r") as file:
    loaded_game = json.load(file, object_hook=custom_decoder)

print(loaded_game.level)  # 3
print(loaded_game.player.name)  # Alice
```

### **Complete Notes on Pickling in Python**

Pickling in Python is the process of serializing objects into a byte stream that can be stored on disk or transmitted over a network. The serialized object can later be deserialized (unpickled) to reconstruct the original Python object in memory.

---

### **What is Pickling?**
- **Pickling**: Converting a Python object into a byte stream using the `pickle` module.
- **Unpickling**: Converting a byte stream back into a Python object.

---

### **Why Use Pickling?**
1. **Data Persistence**: Save objects like dictionaries, lists, or custom class instances to files.
2. **Data Transfer**: Share Python objects over a network.
3. **Caching**: Store pre-computed results for reuse without recalculating.

---

### **How Pickling Works?**
1. The `pickle` module provides methods like `dump()` and `load()` to handle serialization and deserialization.
2. Python objects are encoded into a binary format that can be written to files or transmitted.
3. Only objects supported by the `pickle` module can be serialized.

---

### **Pickle Module Methods**

| **Method**                | **Description**                                                                 |
|----------------------------|---------------------------------------------------------------------------------|
| `pickle.dump(obj, file)`   | Serialize `obj` to the file-like object `file`.                                 |
| `pickle.load(file)`        | Deserialize the file-like object `file` into a Python object.                  |
| `pickle.dumps(obj)`        | Serialize `obj` into a bytes object.                                           |
| `pickle.loads(bytes)`      | Deserialize `bytes` into a Python object.                                      |

---

### **Basic Example of Pickling**

```python
import pickle
```

### **Supported and Unsupported Objects**
#### **Supported Objects**
- Built-in data types: integers, floats, strings, tuples, lists, dictionaries, etc.
- Functions (but not lambda functions).

#### **Unsupported Objects**
- File handles, sockets, database connections, etc.
- Lambda functions (due to their anonymous nature).
- Objects dependent on external resources (e.g., threads, OS-level handles).

### **Customizing Pickling for Custom Objects**

#### **Using `__getstate__` and `__setstate__`**
- Use `__getstate__()` to customize what is serialized.
- Use `__setstate__()` to define how to reconstruct the object during deserialization.

**Example:**
```python
import pickle

class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def __getstate__(self):
        state = self.__dict__.copy()
        state['age'] += 1  # Modify data before pickling
        return state

    def __setstate__(self, state):
        state['age'] -= 1  # Revert modification during unpickling
        self.__dict__ = state

# Serialize and Deserialize
person = Person("Alice", 30)
with open('person.pkl', 'wb') as file:
    pickle.dump(person, file)

with open('person.pkl', 'rb') as file:
    loaded_person = pickle.load(file)

print(loaded_person.age)  # 30
```

Pickle is often used to serialize data sent between processes or over sockets.

```python
import socket

```

### **Security Risks of Pickling**

1. **Code Execution**: Pickle can execute arbitrary code during deserialization, making it insecure for loading data from untrusted sources.
2. **Mitigation**: Always validate and sanitize untrusted input. Avoid using `pickle` for web applications or data from unknown sources.

---

### **Pickle Protocol Versions**

- **Protocol 0**: ASCII format (backward compatible with Python 2.x).
- **Protocol 1**: Binary format (backward compatible with Python 2.x).
- **Protocol 2**: Added for Python 2.3, includes new features like optimized representation for new-style classes.
- **Protocol 3**: Added in Python 3.0, includes support for binary data (not backward compatible with Python 2.x).
- **Protocol 4**: Added in Python 3.4, supports large data.
- **Protocol 5**: Added in Python 3.8, optimizes out-of-band data.

**Choosing a Protocol:**
- Use `protocol=pickle.HIGHEST_PROTOCOL` to select the latest protocol automatically.

---

### **Pickle Alternatives**

1. **`json` Module**:
   - Human-readable and cross-language compatible.
   - Doesn't support complex Python objects.

2. **`marshal` Module**:
   - Faster than `pickle` but limited to basic Python objects.
   - Used internally by Python (e.g., for `.pyc` files).

3. **`dill` Library**:
   - Extends `pickle` to support more object types like lambdas and closures.

4. **`joblib` Library**:
   - Optimized for large data like NumPy arrays.

---

### **Best Practices for Pickling**
1. **Avoid Untrusted Data**: Only unpickle data from trusted sources.
2. **Version Control**: Ensure compatibility when upgrading Python versions.
3. **Use the Latest Protocol**: For better performance and feature support, use `protocol=pickle.HIGHEST_PROTOCOL`.

### **Common Errors**

### Detailed Notes on **Introduction to OOP (Object-Oriented Programming) and Class & Object in Python**

---

### **What is Object-Oriented Programming (OOP)?**

**Object-Oriented Programming (OOP)** is a programming paradigm based on the concept of **"objects"**, which can contain data (attributes) and code (methods). OOP enables better organization of code, reuse of functionalities, and modularity.

---

### **Core Principles of OOP**

1. **Encapsulation**:
   - Bundling data (attributes) and methods (functions) that operate on the data into a single unit (class).
   - Controls access to the data using access modifiers (`public`, `protected`, `private`).

2. **Abstraction**:
   - Hiding the complex implementation details and exposing only the essential features of an object.
   - Achieved using abstract classes and interfaces.

3. **Inheritance**:
   - Allows one class to inherit attributes and methods from another class.
   - Enables code reuse and establishes a hierarchy.

4. **Polymorphism**:
   - The ability of different classes to respond to the same method call in different ways.
   - Achieved through method overloading, method overriding, and operator overloading.

---

### **Advantages of OOP**

- **Modularity**: Code is divided into reusable and independent objects.
- **Code Reusability**: Classes and methods can be reused in other parts of the program.
- **Improved Maintainability**: Changes in one part of the code can be made without affecting others.
- **Real-World Mapping**: Models real-world entities as objects.
- **Flexibility**: Allows adding new features easily using inheritance and polymorphism.

---

### **Class and Object in Python**

#### **What is a Class?**
- A **class** is a blueprint for creating objects.
- It defines the properties (attributes) and behaviors (methods) that the objects of the class will have.

#### **Syntax for Defining a Class**
```python
class ClassName:
    # Class attributes and methods
    pass
```

#### **What is an Object?**
- An **object** is an instance of a class.
- It is a real-world entity that has attributes (data) and behaviors (methods).

#### **Creating an Object**
```python
# Define a class
class Person:
    pass

# Create an object of the class
person1 = Person()
```

---

### **Example: Class and Object**

```python
class Person:
    # Constructor
    def __init__(self, name, age):
        self.name = name  # Attribute
        self.age = age    # Attribute

    # Method
    def greet(self):
        return f"Hello, my name is {self.name} and I am {self.age} years old."

# Create objects
person1 = Person("Alice", 25)
person2 = Person("Bob", 30)

# Access attributes and methods
print(person1.greet())  # Output: Hello, my name is Alice and I am 25 years old.
print(person2.greet())  # Output: Hello, my name is Bob and I am 30 years old.
```

---

### **Key Components of a Class**

1. **Attributes**:
   - Variables that belong to a class.
   - Define the properties of the object.
   - Declared inside the `__init__()` method or directly within the class for class-level attributes.

2. **Methods**:
   - Functions defined inside a class to perform actions on the object’s data.

3. **Constructor (`__init__` Method)**:
   - A special method that gets called automatically when an object is created.
   - Used to initialize the object's attributes.

4. **Self Parameter**:
   - Represents the instance of the class.
   - Used to access attributes and methods of the class.

---

### **Types of Attributes**

1. **Instance Attributes**:
   - Defined inside the constructor (`__init__` method).
   - Unique to each object.

   Example:
   ```python
   class Car:
       def __init__(self, color):
           self.color = color

   car1 = Car("Red")
   car2 = Car("Blue")
   print(car1.color)  # Output: Red
   print(car2.color)  # Output: Blue
   ```

2. **Class Attributes**:
   - Shared across all instances of the class.
   - Declared directly in the class body.

   Example:
   ```python
   class Car:
       wheels = 4  # Class attribute

   car1 = Car()
   car2 = Car()
   print(car1.wheels)  # Output: 4
   print(car2.wheels)  # Output: 4
   ```

---

### **Types of Methods**

1. **Instance Methods**:
   - Operate on instance attributes.
   - Require the `self` parameter.

2. **Class Methods**:
   - Operate on class attributes.
   - Defined using the `@classmethod` decorator.
   - Use `cls` as the first parameter.

3. **Static Methods**:
   - Do not operate on instance or class attributes.
   - Defined using the `@staticmethod` decorator.
   - Behave like regular functions but belong to the class.

   Example:
   ```python
   class Math:
       @staticmethod
       def add(x, y):
           return x + y
   print(Math.add(5, 10))  # Output: 15
   ```

---

### **Accessing Class Components**

1. **Accessing Attributes**:
   ```python
   object.attribute_name
   ```

2. **Calling Methods**:
   ```python
   object.method_name()
   ```

---

### **Key Terms**

- **Encapsulation**: Binding data and methods together.
- **Abstraction**: Hiding unnecessary details.
- **Constructor**: A special method for initializing objects.
- **Inheritance**: Reusing code by deriving new classes from existing ones.
- **Polymorphism**: Using the same interface for different types of objects.

---

### **Important Notes**
- A class can have multiple objects, each with its own data.
- Objects share class-level attributes but have their own instance-level attributes.
- OOP enables you to write scalable, reusable, and modular code.

### Detailed and Complete Notes on Class Diagrams in Python

---

### **What is a Class Diagram?**

A **class diagram** is a type of UML (Unified Modeling Language) diagram that represents the **structure** of a system by modeling its classes, attributes, methods, and relationships. It is widely used in **Object-Oriented Design (OOD)** to plan and visualize the architecture of a system.

In Python, while we don't directly use class diagrams, they help us design and understand the relationships and interactions between classes before coding.

---

### **Key Components of a Class Diagram**

1. **Classes**:
   - Represented by rectangles divided into three sections:
     1. **Class Name**: The name of the class.
     2. **Attributes**: The properties or data fields of the class.
     3. **Methods**: The functions or behaviors of the class.

2. **Attributes**:
   - Define the data stored in the class.
   - Typically include variables and constants.

3. **Methods**:
   - Define the behavior or functionality of the class.
   - Represent the actions that objects of the class can perform.

4. **Relationships**:
   - Represent how classes are connected and interact.
   - Types of relationships:
     - **Association**: General relationship between classes.
     - **Aggregation**: "Has-a" relationship where one class contains another but can exist independently.
     - **Composition**: Stronger "Has-a" relationship where one class contains another and cannot exist independently.
     - **Inheritance**: "Is-a" relationship between parent and child classes.
     - **Dependency**: Temporary usage relationship where one class depends on another for a task.

---

### **Class Diagram Notation**

1. **Class Structure**:
   ```
   +----------------+
   | ClassName      |   <-- Class Name
   +----------------+
   | Attribute1     |   <-- Attributes
   | Attribute2     |
   +----------------+
   | Method1()      |   <-- Methods
   | Method2()      |
   +----------------+
   ```

2. **Visibility Modifiers**:
   - `+` (Public): Accessible from anywhere.
   - `-` (Private): Accessible only within the class.
   - `#` (Protected): Accessible within the class and its subclasses.

3. **Relationships Notations**:
   - **Association**: Solid line.
   - **Inheritance**: Solid line with a hollow arrow pointing to the parent.
   - **Aggregation**: Solid line with a diamond near the container class.
   - **Composition**: Solid line with a filled diamond near the container class.
   - **Dependency**: Dashed line with an arrow.

---

### **Steps to Create a Class Diagram**

1. **Identify the Classes**:
   - Determine the entities in the system that will become classes.

2. **Define Attributes and Methods**:
   - List the data (attributes) and behaviors (methods) each class will have.

3. **Establish Relationships**:
   - Define how the classes interact with one another.

4. **Specify Access Modifiers**:
   - Decide whether attributes and methods should be public, private, or protected.

5. **Draw the Diagram**:
   - Use UML tools or draw it manually to represent the class structure.

---

### **Creating Python Code from a Class Diagram**

Here’s the Python code based on the above class diagram:

```python
class Book:
    def __init__(self, title, author, isbn):
        self.title = title
        self.author = author
        self.isbn = isbn
        self.is_borrowed = False

    def borrow(self):
        if not self.is_borrowed:
            self.is_borrowed = True
            print(f"'{self.title}' has been borrowed.")
        else:
            print(f"'{self.title}' is already borrowed.")

    def return_book(self):
        if self.is_borrowed:
            self.is_borrowed = False
            print(f"'{self.title}' has been returned.")
        else:
            print(f"'{self.title}' was not borrowed.")

class Member:
    def __init__(self, name, member_id):
        self.name = name
        self.member_id = member_id
        self.borrowed_books = []

    def borrow_book(self, book):
        if book not in self.borrowed_books:
            self.borrowed_books.append(book)
            book.borrow()
        else:
            print(f"{self.name} already borrowed '{book.title}'.")

    def return_book(self, book):
        if book in self.borrowed_books:
            self.borrowed_books.remove(book)
            book.return_book()
        else:
            print(f"{self.name} did not borrow '{book.title}'.")

class Library:
    def __init__(self):
        self.books = []
        self.members = []

    def add_book(self, book):
        self.books.append(book)
        print(f"'{book.title}' has been added to the library.")

    def register_member(self, member):
        self.members.append(member)
        print(f"Member '{member.name}' has been registered.")

    def issue_book(self, member, book):
        if book in self.books and book not in member.borrowed_books:
            member.borrow_book(book)
        else:
            print(f"'{book.title}' is not available or already issued.")
```

### Detailed and Complete Notes on Magic Methods in Python

---

### **What are Magic Methods?**

Magic methods, also known as **dunder methods** (short for "double underscore"), are special methods in Python that start and end with double underscores (e.g., `__init__`, `__str__`). These methods are predefined in Python and provide a way to customize the behavior of objects, enabling operator overloading, object initialization, attribute access, and more.

Magic methods are not called explicitly; they are invoked implicitly by Python in response to specific operations or constructs.

---

### **Categories of Magic Methods**

1. **Initialization and Representation Methods**
2. **Arithmetic and Comparison Operators**
3. **Container and Sequence Emulation**
4. **Attribute Access and Management**
5. **Context Management**
6. **Other Special Methods**

---

### **1. Initialization and Representation Methods**

#### `__init__(self, ...)`
- Called when an object is instantiated.
- Used to initialize attributes.

Example:
```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

p = Person("Alice", 25)
```

#### `__str__(self)`
- Defines the string representation of an object (used by `print()` and `str()`).

Example:
```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def __str__(self):
        return f"{self.name}, {self.age} years old"

p = Person("Alice", 25)
print(p)  # Output: Alice, 25 years old
```

#### `__repr__(self)`
- Defines an official string representation of an object (used by `repr()`).
- Should be unambiguous and ideally recreate the object.

Example:
```python
class Person:
    def __repr__(self):
        return f"Person('{self.name}', {self.age})"
```

#### `__del__(self)`
- Called when an object is deleted (destructor).

Example:
```python
class Person:
    def __del__(self):
        print(f"{self.name} is deleted")
```

---

### **2. Arithmetic and Comparison Operators**

#### Arithmetic Magic Methods
These methods define behavior for arithmetic operations.

| Operator | Magic Method    |
|----------|-----------------|
| `+`      | `__add__`       |
| `-`      | `__sub__`       |
| `*`      | `__mul__`       |
| `/`      | `__truediv__`   |
| `//`     | `__floordiv__`  |
| `%`      | `__mod__`       |
| `**`     | `__pow__`       |

Example:
```python
class Number:
    def __init__(self, value):
        self.value = value

    def __add__(self, other):
        return Number(self.value + other.value)

n1 = Number(10)
n2 = Number(20)
result = n1 + n2
print(result.value)  # Output: 30
```

#### Comparison Magic Methods
These methods define behavior for comparison operators.

| Operator | Magic Method    |
|----------|-----------------|
| `<`      | `__lt__`        |
| `<=`     | `__le__`        |
| `>`      | `__gt__`        |
| `>=`     | `__ge__`        |
| `==`     | `__eq__`        |
| `!=`     | `__ne__`        |

Example:
```python
class Number:
    def __init__(self, value):
        self.value = value

    def __eq__(self, other):
        return self.value == other.value

n1 = Number(10)
n2 = Number(10)
print(n1 == n2)  # Output: True
```

---

### **3. Container and Sequence Emulation**

#### Container Magic Methods
- `__getitem__(self, key)`: Access an item.
- `__setitem__(self, key, value)`: Set an item.
- `__delitem__(self, key)`: Delete an item.

Example:
```python
class CustomList:
    def __init__(self):
        self.items = []

    def __getitem__(self, index):
        return self.items[index]

    def __setitem__(self, index, value):
        self.items[index] = value

    def __delitem__(self, index):
        del self.items[index]
```

#### Sequence Magic Methods
- `__len__(self)`: Returns the length.
- `__contains__(self, item)`: Checks for membership (`in` operator).

Example:
```python
class CustomList:
    def __init__(self, items):
        self.items = items

    def __len__(self):
        return len(self.items)

    def __contains__(self, item):
        return item in self.items

cl = CustomList([1, 2, 3])
print(len(cl))      # Output: 3
print(2 in cl)      # Output: True
```

---

### **4. Attribute Access and Management**

#### `__getattr__(self, name)`
- Called when an attribute is not found in the object.

#### `__setattr__(self, name, value)`
- Called when setting an attribute.

#### `__delattr__(self, name)`
- Called when deleting an attribute.

Example:
```python
class Person:
    def __setattr__(self, name, value):
        self.__dict__[name] = value.upper()

p = Person()
p.name = "alice"

```

---

### **5. Context Management**

Magic methods used for context management (`with` statement).

- `__enter__(self)`: Enter the runtime context.
- `__exit__(self, exc_type, exc_value, traceback)`: Exit the runtime context.

Example:
```python
class FileManager:
    def __init__(self, filename):
        self.filename = filename

    def __enter__(self):
        self.file = open(self.filename, 'w')
        return self.file

    def __exit__(self, exc_type, exc_value, traceback):
        self.file.close()

with FileManager('example.txt') as f:
    f.write('Hello, world!')
```

---

### **6. Other Special Methods**

- `__call__(self, ...)`: Makes an object callable like a function.
- `__hash__(self)`: Returns the hash value of an object.
- `__bool__(self)`: Defines the boolean value of an object.

Example:
```python
class CallableClass:
    def __call__(self, x):
        return x * 2

cc = CallableClass()
print(cc(5))  # Output: 10
```

---

### **Practical Applications of Magic Methods**

1. **Operator Overloading**: Customize operators for your classes.
2. **Custom Data Structures**: Mimic lists, dictionaries, etc.
3. **Advanced Context Management**: Manage resources like files or database connections.
4. **Readable Objects**: Enhance debugging by customizing `__str__` and `__repr__`.
5. **Dynamic Behavior**: Implement features like attribute access and callable objects.

---

### **Tips and Best Practices**

1. **Use Magic Methods Wisely**: Don't overuse them; they can make code harder to understand.
2. **Follow Pythonic Principles**: Ensure behavior aligns with Python conventions.
3. **Document Your Methods**: Clearly explain overridden magic methods for maintainability.

---

### **Summary**

Magic methods are an advanced feature of Python that enable you to enhance and customize the behavior of objects. They are essential for building robust, Pythonic classes that interact seamlessly with Python’s built-in operations. With proper understanding and usage, they can significantly improve the elegance and functionality of your code.

### Detailed and Complete Notes on Constructors in Python

---

### **What is a Constructor?**

A **constructor** is a special method in Python used to initialize an object’s state when it is created. It allows you to define and set up instance variables and perform other necessary operations at the time of object creation.

In Python, the constructor is defined using the `__init__()` magic method.

---

### **Purpose of a Constructor**

1. **Initialization**: Assign default or user-defined values to instance variables.
2. **Encapsulation**: Group related data and functionality into a single entity (the object).
3. **Automation**: Perform setup tasks like opening files, initializing connections, etc.

---

### **How to Define a Constructor**

The constructor is defined using the `__init__()` method, which is automatically called when a new object is instantiated.

#### Syntax:
```python
class ClassName:
    def __init__(self, parameters):
        # Initialization code here
```

#### Example:
```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

p = Person("Alice", 25)
print(p.name)  # Output: Alice
print(p.age)   # Output: 25
```

---

### **Types of Constructors in Python**

1. **Default Constructor**
2. **Parameterized Constructor**

---

#### 1. **Default Constructor**

A constructor with no arguments (except `self`). It is used when the initial values of instance variables are predefined or do not depend on user input.

Example:
```python
class DefaultExample:
    def __init__(self):
        self.value = 42

obj = DefaultExample()
print(obj.value)  # Output: 42
```

---

#### 2. **Parameterized Constructor**

A constructor that takes arguments to initialize instance variables with user-defined values.

Example:
```python
class ParameterizedExample:
    def __init__(self, data):
        self.data = data

obj = ParameterizedExample("Hello, World!")
print(obj.data)  # Output: Hello, World!
```

---

### **Key Points About Constructors**

1. **`self` Parameter**:
   - Represents the instance of the class.
   - Used to access attributes and methods within the class.
   - Must be the first parameter in `__init__()` but is not explicitly passed during object creation.

2. **Automatic Invocation**:
   - The constructor is called automatically when an object is created.
   - You do not need to call it explicitly.

3. **Overriding Constructors**:
   - A class can have only one constructor. Defining multiple `__init__()` methods will override the previous one.

---

### **Examples of Constructor Use Cases**

#### Example 1: Setting Default Values
```python
class Person:
    def __init__(self, name="Unknown", age=0):
        self.name = name
        self.age = age

p1 = Person()
p2 = Person("Alice", 25)
print(p1.name, p1.age)  # Output: Unknown 0
print(p2.name, p2.age)  # Output: Alice 25
```

#### Example 2: Incrementing a Class Counter
```python
class Person:
    count = 0  # Class variable

    def __init__(self, name):
        self.name = name
        Person.count += 1

p1 = Person("Alice")
p2 = Person("Bob")
print(Person.count)  # Output: 2
```

#### Example 3: Validating Input
```python
class Rectangle:
    def __init__(self, length, width):
        if length <= 0 or width <= 0:
            raise ValueError("Length and width must be positive numbers.")
        self.length = length
        self.width = width

rect = Rectangle(10, 5)
print(rect.length, rect.width)  # Output: 10 5
```

---

### **Constructor in Inheritance**

In Python, constructors of parent and child classes interact in inheritance scenarios. If a child class does not have its own constructor, the parent class constructor is automatically called.

#### Example 1: Using Parent Constructor
```python
class Parent:
    def __init__(self, name):
        self.name = name

class Child(Parent):
    pass

c = Child("Alice")
print(c.name)  # Output: Alice
```

#### Example 2: Overriding Parent Constructor
```python
class Parent:
    def __init__(self, name):
        self.name = name

class Child(Parent):
    def __init__(self, name, age):
        super().__init__(name)  # Call parent constructor
        self.age = age

c = Child("Alice", 25)
print(c.name, c.age)  # Output: Alice 25
```

---

### **Multiple Constructors Using Default Parameters**

Python does not support multiple constructors directly, but you can simulate them using default parameters.

#### Example:
```python
class Person:
    def __init__(self, name=None, age=None):
        if name and age:
            self.name = name
            self.age = age
        elif name:
            self.name = name
            self.age = "Unknown"
        else:
            self.name = "Unknown"
            self.age = "Unknown"

p1 = Person()
p2 = Person("Alice")
```

### **Destructor vs Constructor**

Example of Destructor:

- **Single Constructor**: A class can have only one `__init__` method.
- **Automatic Invocation**: Called automatically when creating an object.

### Complete and Detailed Notes on `self` in Python

---

### **What is `self` in Python?**

- In Python, `self` is a **conventionally used variable name** that refers to the **current instance of a class**.
- It is used to access attributes and methods associated with the object that is being manipulated.
- It must be explicitly declared as the **first parameter** of instance methods in a class.

---

### **Why Do We Use `self`?**

1. **Access Instance Variables**: To distinguish between instance variables and local variables.
2. **Bind Methods to Instances**: `self` ensures that the methods act on the specific object of the class.
3. **Dynamic Binding**: It helps Python determine which object to apply an operation to, even when multiple objects exist.

---

### **Key Points About `self`**

1. **Not a Keyword**: `self` is not a reserved keyword in Python; it is merely a naming convention. You can use any valid variable name, but using `self` is strongly recommended for readability and consistency.
   
2. **Mandatory in Instance Methods**: You must include `self` as the first parameter of instance methods, even though you don’t pass it explicitly while calling the method.

3. **Not Used for Static or Class Methods**: 
   - **Static Methods**: They do not take `self` because they do not operate on an instance.
   - **Class Methods**: They take `cls` instead of `self` as their first parameter.

---

### **How `self` Works**

When you call a method on an object, Python automatically passes the object itself as the first argument to the method. This is why you need `self` in the method definition.

#### Example:
```python
class MyClass:
    def my_method(self):  # 'self' refers to the instance of the class
        print("Method called from", self)

obj = MyClass()
obj.my_method()
```

Output:
```
Method called from <__main__.MyClass object at 0x...>
```

---

### **Using `self` to Access Attributes**

#### Example:
```python
class Person:
    def __init__(self, name, age):
        self.name = name  # Assign instance variable 'name'
        self.age = age    # Assign instance variable 'age'

    def display(self):
        print(f"My name is {self.name} and I am {self.age} years old.")

# Creating an object
person1 = Person("Alice", 25)
person1.display()
```

Output:
```
My name is Alice and I am 25 years old.
```

---

### **Using `self` to Call Other Methods**

#### Example:
```python
class Calculator:
    def add(self, a, b):
        return a + b

    def multiply(self, a, b):
        result = self.add(a, b) * b  # Calling another method using 'self'
        return result

calc = Calculator()
print(calc.multiply(2, 3))
```

Output:
```
15
```

---

### **`self` in Multiple Objects**

Each object has its own copy of instance variables, and `self` ensures that each object operates on its own data.

#### Example:
```python
class Counter:
    def __init__(self):
        self.count = 0

    def increment(self):
        self.count += 1

c1 = Counter()
c2 = Counter()

c1.increment()
c1.increment()
c2.increment()

print(c1.count)  # Output: 2
print(c2.count)  # Output: 1
```

---

### **Changing `self` Variable Name**

Although it is not recommended, you can use another name instead of `self`.

#### Example:
```python

    def display(this):  # Using 'this' instead of 'self'
        print("This works too!")

obj.display()
```

Output:
```

```

However, sticking to `self` is advised for clarity and convention.

---

### **`self` with Inheritance**

When working with inheritance, `self` refers to the instance of the child class when accessing attributes or methods.

#### Example:
```python
class Parent:
    def greet(self):
        print("Hello from Parent!")

class Child(Parent):
    def greet(self):
        print("Hello from Child!")
        super().greet()  # Using super() to call parent method

c = Child()
c.greet()
```

Output:
```
Hello from Child!
Hello from Parent!
```

---

### **Difference Between `self` and `cls`**

| Feature             | `self`                          | `cls`                          |
|---------------------|----------------------------------|--------------------------------|
| Usage              | Refers to the instance of a class | Refers to the class itself     |
| Used In            | Instance methods                 | Class methods                  |
| Passed Automatically | Yes                             | Yes                            |

#### Example of `cls`:
```python
class MyClass:
    class_variable = "Class Level"

    @classmethod
    def show(cls):
        print(f"Accessing {cls.class_variable} using cls")

MyClass.show()
```

Output:
```
Accessing Class Level using cls
```

---

### **Common Mistakes with `self`**

1. **Forgetting to Include `self` in Method Definitions**:
   ```python
   class MyClass:
       def my_method():  # Missing 'self'
           print("Error!")
   ```

   This will raise an error:  
   ```
   TypeError: my_method() takes 0 positional arguments but 1 was given
   ```

2. **Using `self` Without Understanding Its Scope**:
   - Attempting to use `self` in a static method will result in an error because `self` is not passed automatically.

3. **Not Using `self` to Access Instance Variables**:
   ```python
   class MyClass:
       def __init__(self, value):
           value = value  # This won't assign to instance variable
   ```

   Corrected Version:
   ```python
   self.value = value
   ```

### **Advanced Use Cases of `self`**

#### Example 1: Dynamic Attribute Assignment
```python
class DynamicAttributes:
    def __init__(self, **kwargs):
        for key, value in kwargs.items():
            setattr(self, key, value)

```

Output:

```python
        return self

    def multiply(self, num):
        self.value *= num
        return self

```

   - Access instance variables.
   - Call other instance methods.
   - Differentiate between class and instance scope.
3. It must be explicitly defined as the first parameter of instance methods.
4. `self` is automatically passed to instance methods when they are called.
5. Stick to the naming convention of using `self` for clarity and consistency.

By understanding `self`, you can master object-oriented programming concepts in Python and build robust and reusable code.

````python
class Fraction:
    
    # parameterized constructor => ka mtlb yee hai ham ek aisa constructor bna rhe hai jisse input ek zarurat hote hai
    def __init__(self,x,y):
        self.num = x
        self.den = y
    
    # second magic method
    def __str__(self):
        return f"{self.num}/{self.den}"
    
    # second magic method
    def __add__(self,other):
       new_num = (self.num * other.den)+ (other.num * self.den)
       new_den = self.den * other.den
       
       return f"{new_num}/{new_den}"
   
    def __sub__(self,other):
       new_num = (self.num * other.den) - (other.num * self.den)
       new_den = self.den * other.den
       
       return f"{new_num}/{new_den}"
   
    def __mul__(self,other):
       new_num = self.num * other.num
       new_den = self.den * other.den
       
       return f"{new_num}/{new_den}"
     
    def __truediv__(self,other):
       new_num = self.num * other.den
       new_den = self.den * other.num
       
       return f"{new_num}/{new_den}"
   
    def convert_to_decimal(self):
        return self.num/self.den
````

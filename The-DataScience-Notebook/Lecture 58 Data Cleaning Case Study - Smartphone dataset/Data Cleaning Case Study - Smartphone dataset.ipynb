<center> Completeness

```python
clean_smartphone.tail(8)
```

```python
# Dropping NaN values
clean_smartphone.dropna(inplace=True)
```

<center> Validity

```python
# Clean price column

clean_smartphone['Price'] = clean_smartphone['Price'].str.replace(
    '₹', '', regex=False).str.replace(',', '', regex=False)
```

```python
clean_smartphone['Price']
```

```python
# Incorrect dtype of price and spec_score

clean_smartphone['Price'] = clean_smartphone['Price'].astype('Int64')
clean_smartphone['Spec_Score'] = clean_smartphone['Spec_Score'].astype('Int64')
```

<center> Consistency

```python
# Remove RAM/ROM info inside brackets in Product_name columns

clean_smartphone['Product_Name'] = clean_smartphone['Product_Name'].str.replace(
    r'\s*\([^)]*[Gg][Bb][^)]*\)', '', regex=True).str.strip()
```

```python
# Standardize brand spelling (OPPO, etc.)

clean_smartphone['Product_Name'] = clean_smartphone['Product_Name'].str.replace(
    r'(?i)^oppo', 'OPPO', regex=True)
```

```python
clean_smartphone.info()
```

<center> Tidiness issues

```python
# ============================================================================
# TIDINESS ISSUES - ROBUST FEATURE EXTRACTION FROM SPECIFICATIONS
# ============================================================================

# Create a copy of Specifications for easier access
clean_smartphone['spec_text'] = clean_smartphone['Specifications'].fillna('')
```

```python
clean_smartphone['spec_text'].str.contains(
    r'Dual\s*Sim', case=False, regex=True).astype(int)
```

```python
# ===== CONNECTIVITY FEATURES =====
clean_smartphone['has_dual_sim'] = clean_smartphone['spec_text'].str.contains(
    r'Dual\s*Sim', case=False, regex=True).astype(int)
clean_smartphone['has_3g'] = clean_smartphone['spec_text'].str.contains(
    r'\b3G\b', case=False, regex=True).astype(int)
clean_smartphone['has_4g'] = clean_smartphone['spec_text'].str.contains(
    r'\b4G\b', case=False, regex=True).astype(int)
clean_smartphone['has_5g'] = clean_smartphone['spec_text'].str.contains(
    r'\b5G\b', case=False, regex=True).astype(int)
clean_smartphone['has_nfc'] = clean_smartphone['spec_text'].str.contains(
    r'\bNFC\b', case=False, regex=True).astype(int)
clean_smartphone['has_ir_blaster'] = clean_smartphone['spec_text'].str.contains(
    r'IR\s*Blaster', case=False, regex=True).astype(int)
```

```python
clean_smartphone.head(3)
```

```python
# ===== PROCESSOR INFORMATION =====
import re

# Extract processor name - look for pattern before "Processor" keyword or before core info
processor_pattern = r'\|\s*([^|,]+?)(?:,\s*(?:Octa|Quad|Hexa|Deca|Dual|Single|Nine-Cores)\s*Core|\s+Processor)'
proc_match = clean_smartphone['spec_text'].str.extract(
    processor_pattern, flags=re.IGNORECASE)
clean_smartphone['processor_name'] = proc_match[0].str.strip()

# Extract cores
core_pattern = r'(Octa|Quad|Hexa|Deca|Dual|Single)\s*Core'
clean_smartphone['cores'] = clean_smartphone['spec_text'].str.extract(
    core_pattern, flags=re.IGNORECASE)[0]
clean_smartphone['cores'] = clean_smartphone['cores'].str.title() + ' Core'
clean_smartphone.loc[clean_smartphone['cores'] == 'Nan Core', 'cores'] = pd.NA

# Extract CPU speed in GHz
ghz_pattern = r'([0-9.]+)\s*GHz'
clean_smartphone['cpu_speed_ghz'] = clean_smartphone['spec_text'].str.extract(
    ghz_pattern)[0]
clean_smartphone['cpu_speed_ghz'] = pd.to_numeric(
    clean_smartphone['cpu_speed_ghz'], errors='coerce')
```

```python
# ===== RAM AND ROM =====
print("  - Extracting RAM and ROM...")
ram_pattern = r'(\d+)\s*GB\s*RAM'
clean_smartphone['ram_gb'] = clean_smartphone['spec_text'].str.extract(
    ram_pattern, flags=re.IGNORECASE)[0]
clean_smartphone['ram_gb'] = pd.to_numeric(
    clean_smartphone['ram_gb'], errors='coerce')

rom_pattern = r'(\d+)\s*GB\s*(?:inbuilt|internal|storage)'
clean_smartphone['rom_gb'] = clean_smartphone['spec_text'].str.extract(
    rom_pattern, flags=re.IGNORECASE)[0]
clean_smartphone['rom_gb'] = pd.to_numeric(
    clean_smartphone['rom_gb'], errors='coerce')
```

```python
# ===== BATTERY =====
print("  - Extracting battery information...")
battery_pattern = r'(\d+)\s*mAh'
clean_smartphone['battery_capacity_mah'] = clean_smartphone['spec_text'].str.extract(
    battery_pattern)[0]
clean_smartphone['battery_capacity_mah'] = pd.to_numeric(
    clean_smartphone['battery_capacity_mah'], errors='coerce')

clean_smartphone['fast_charging_available'] = clean_smartphone['spec_text'].str.contains(
    r'Fast\s*Charging', case=False, regex=True).astype(int)

fast_charge_pattern = r'(\d+)W\s*Fast\s*Charging'
clean_smartphone['fast_charging_speed_w'] = clean_smartphone['spec_text'].str.extract(
    fast_charge_pattern, flags=re.IGNORECASE)[0]
clean_smartphone['fast_charging_speed_w'] = pd.to_numeric(
    clean_smartphone['fast_charging_speed_w'], errors='coerce')
```

```python
# ===== DISPLAY =====
print("  - Extracting display information...")
display_size_pattern = r'([0-9.]+)\s*inches'
clean_smartphone['display_size_inches'] = clean_smartphone['spec_text'].str.extract(
    display_size_pattern)[0]
clean_smartphone['display_size_inches'] = pd.to_numeric(
    clean_smartphone['display_size_inches'], errors='coerce')

resolution_pattern = r'(\d{3,4})\s*x\s*(\d{3,4})\s*px'
resolution = clean_smartphone['spec_text'].str.extract(resolution_pattern)
clean_smartphone['resolution_width'] = pd.to_numeric(
    resolution[0], errors='coerce')
clean_smartphone['resolution_height'] = pd.to_numeric(
    resolution[1], errors='coerce')

freq_pattern = r'(\d{2,3})\s*Hz\s*Display'
clean_smartphone['display_frequency_hz'] = clean_smartphone['spec_text'].str.extract(
    freq_pattern)[0]
clean_smartphone['display_frequency_hz'] = pd.to_numeric(
    clean_smartphone['display_frequency_hz'], errors='coerce')
```

```python
# ===== CAMERA =====
print("  - Extracting camera information...")
# Front camera - look for pattern with "Front Camera"
front_pattern = r'(?:&|\|)\s*(\d+)\s*MP\s*Front\s*Camera'
clean_smartphone['front_camera_mp'] = clean_smartphone['spec_text'].str.extract(
    front_pattern, flags=re.IGNORECASE)[0]
clean_smartphone['front_camera_mp'] = pd.to_numeric(
    clean_smartphone['front_camera_mp'], errors='coerce')

# Rear camera - look for first MP value before "&" or "Rear"
rear_pattern = r'(\d+)\s*MP.*?(?:Rear|Camera)'
rear_matches = clean_smartphone['spec_text'].str.extract(
    rear_pattern, flags=re.IGNORECASE)[0]
clean_smartphone['rear_camera_mp'] = pd.to_numeric(
    rear_matches, errors='coerce')
```

```python
# ===== MEMORY CARD =====
print("  - Extracting memory card information...")
# Check if NOT supported
clean_smartphone['card_supported'] = (~clean_smartphone['spec_text'].str.contains(
    'Memory Card Not Supported|Card Not Supported', case=False, regex=True, na=False)).astype(int)
# Set to NA where spec_text is empty
clean_smartphone.loc[clean_smartphone['spec_text']
                     == '', 'card_supported'] = pd.NA
```

```python
# ===== OPERATING SYSTEM =====
print("  - Extracting operating system...")
# Look for Android, iOS, or other OS
os_pattern = r'(Android|iOS|Windows|HarmonyOS)\s*(?:v)?([0-9.]+)?'
os_match = clean_smartphone['spec_text'].str.extract(
    os_pattern, flags=re.IGNORECASE)
clean_smartphone['operating_system'] = os_match[0] + \
    ' v' + os_match[1].fillna('')
clean_smartphone['operating_system'] = clean_smartphone['operating_system'].str.replace(
    ' v$', '', regex=True)
clean_smartphone.loc[clean_smartphone['operating_system']
                     == '', 'operating_system'] = pd.NA
```

```python
# ============================================================================
# DROP UNNECESSARY COLUMNS
# ============================================================================

print("\n[6/7] Dropping unnecessary columns...")
columns_to_drop = ['Serial_No', 'spec_text']
clean_smartphone = clean_smartphone.drop(
    columns=columns_to_drop, errors='ignore')
print("✓ Unnecessary columns dropped")
```

```python
clean_smartphone.info()
```

```python
# ============================================================================
# FINAL ORGANIZATION
# ============================================================================

print("\n[7/7] Final data organization...")
column_order = ['Product_Name', 'Price', 'Rating', 'Spec_Score', 'Specifications',
                'has_dual_sim', 'has_3g', 'has_4g', 'has_5g', 'has_nfc', 'has_ir_blaster',
                'processor_name', 'cores', 'cpu_speed_ghz',
                'ram_gb', 'rom_gb',
                'battery_capacity_mah', 'fast_charging_available', 'fast_charging_speed_w',
                'display_size_inches', 'resolution_width', 'resolution_height', 'display_frequency_hz',
                'front_camera_mp', 'rear_camera_mp',
                'card_supported',
                'operating_system',
                'Product_Link', 'Image_URL']

existing_columns = [
    col for col in column_order if col in clean_smartphone.columns]
clean_smartphone = clean_smartphone[existing_columns]
print("✓ Columns reordered")
```

```python
# ============================================================================
# SAVE AND DISPLAY RESULTS
# ============================================================================

print("\n" + "="*80)
print("Cleaning Complete!")
print("="*80)
print(f"\nFinal Dataset Shape: {clean_smartphone.shape}")
print(f"Columns: {len(clean_smartphone.columns)}")

# Calculate extraction success rates
print("\n" + "="*80)
print("EXTRACTION SUCCESS RATES:")
print("="*80)
total_non_null = len(clean_smartphone) - 8  # Excluding the 8 rows with no data

for col in ['ram_gb', 'rom_gb', 'processor_name', 'cores', 'cpu_speed_ghz', 'battery_capacity_mah',
            'display_size_inches', 'front_camera_mp', 'rear_camera_mp', 'operating_system']:
    if col in clean_smartphone.columns:
        non_null = clean_smartphone[col].notna().sum()
        percentage = (non_null / total_non_null) * 100
        print(f"{col:30s}: {non_null:4d}/{total_non_null:4d} ({percentage:5.1f}%)")

# Display missing value summary
print("\n" + "="*80)
print("Missing Values Summary:")
print("="*80)
missing_summary = clean_smartphone.isnull().sum()
missing_summary = missing_summary[missing_summary > 0].sort_values(
    ascending=False)
print(missing_summary)

# Save cleaned dataset
output_filename = 'smartprix_mobiles_cleaned.csv'
clean_smartphone.to_csv(output_filename, index=False)
print(f"\n✓ Cleaned dataset saved as: {output_filename}")

# Display first few rows
print("\n" + "="*80)
print("Preview of Cleaned Data (first 3 rows):")
print("="*80)
print(clean_smartphone.head(3).to_string())
```

```python
# Problem 1: Remove "for Galaxy" from Snapdragon 8 Elite processor names

# Load cleaned data
df = pd.read_csv('smartprix_mobiles_cleaned.csv')

print("Before fixing:")
print(df[df['processor_name'].str.contains('for Galaxy',
      case=False, na=False)]['processor_name'].unique())

# Fix the processor name
df['processor_name'] = df['processor_name'].str.replace(
    r'\s*for\s+Galaxy', '', case=False, regex=True)

print("\nAfter fixing:")
print(df[df['processor_name'].str.contains('Snapdragon 8 Elite',
      case=False, na=False)]['processor_name'].unique())
print(f"✓ Fixed 'for Galaxy' issue")
```

```python
# Problem 2-3: Fix wrong processor names (MHz/Hz values)

print("Processors with MHz/Hz:")
mhz_processors = df[df['processor_name'].str.contains(
    r'MHz|Hz', case=False, na=False, regex=True)]
print(f"Count: {len(mhz_processors)}")
print(f"Indices: {mhz_processors.index.tolist()}")
print(f"Values: {mhz_processors['processor_name'].unique()}")

# Strategy: Extract processor name from Specifications for these rows


def extract_processor_from_spec(spec_text):
    """Extract processor name more carefully"""
    if pd.isna(spec_text):
        return np.nan

    # Split by pipe
    parts = str(spec_text).split('|')
    if len(parts) < 2:
        return np.nan

    processor_part = parts[1].strip()

    # Remove "Processor" word
    processor_part = processor_part.replace('Processor', '').strip()

    # Remove core info and GHz info
    # Extract everything before comma or "Core"
    import re
    match = re.search(
        r'^([^,]+?)(?:,|\s+(?:Octa|Quad|Hexa|Deca|Dual|Single)\s*Core)', processor_part, re.IGNORECASE)
    if match:
        return match.group(1).strip()

    # If no match, return the first part before comma
    if ',' in processor_part:
        return processor_part.split(',')[0].strip()

    return processor_part


# Fix rows with MHz/Hz in processor name
for idx in mhz_processors.index:
    spec = df.loc[idx, 'Specifications']
    new_processor = extract_processor_from_spec(spec)

    # If the new processor still has MHz, set to NaN
    if pd.notna(new_processor) and ('MHz' in new_processor or 'Hz' in new_processor):
        df.loc[idx, 'processor_name'] = np.nan
    else:
        df.loc[idx, 'processor_name'] = new_processor

print("\n✓ Fixed MHz/Hz in processor names")
print(
    f"After fix, remaining MHz/Hz processors: {df['processor_name'].str.contains('MHz|Hz', case=False, na=False, regex=True).sum()}")
```

```python
# Problem 3: Extract problematic processor names (Unisoc, MTK, Helio, etc.)

# List of processor patterns to look for
processor_patterns = [
    'Unisoc T127', 'Unisoc T107', 'Unisoc 6533G', 'Unisoc 6531F',
    'MTK MT6739', 'MTK6739', 'Helio 6739', 'AMD'
]

print("Extracting missing processor names...")

for pattern in processor_patterns:
    # Find phones with this processor in specs
    has_pattern = df['Specifications'].str.contains(
        pattern, case=False, na=False, regex=False)

    # Check if processor_name is missing or doesn't contain the pattern
    needs_fix = has_pattern & (df['processor_name'].isna() |
                               ~df['processor_name'].str.contains(pattern, case=False, na=False, regex=False))

    if needs_fix.sum() > 0:
        print(f"\n{pattern}: {needs_fix.sum()} phones need fixing")
        print(f"  Indices: {df[needs_fix].index.tolist()}")

        # Set the processor name
        df.loc[needs_fix, 'processor_name'] = pattern

print("\n✓ Extracted problematic processor names")
```

```python
# Problem 5: Add "Nine Core" to cores extraction

print("Checking Nine Core processors...")

# Find phones with Nine Core in specifications
nine_core_specs = df['Specifications'].str.contains(
    r'Nine\s*Core', case=False, na=False, regex=True)

if nine_core_specs.sum() > 0:
    print(f"Found {nine_core_specs.sum()} phones with Nine Core in specs")
    print(f"Indices: {df[nine_core_specs].index.tolist()}")

    # Fix cores column
    for idx in df[nine_core_specs].index:
        df.loc[idx, 'cores'] = 'Nine Core'

    print("✓ Fixed Nine Core in cores column")
else:
    print("No Nine Core processors found")
```

```python
# Problem 7: Convert MHz values to GHz in specifications

import re
print("Checking for MHz values in specifications that should be GHz...")

# Find phones with MHz in specs (not GHz)
mhz_in_specs = df['Specifications'].str.contains(
    r'\d+\.?\d*\s*MHz', case=False, na=False, regex=True)
print(f"Phones with MHz in specs: {mhz_in_specs.sum()}")

# Extract MHz values and convert to GHz

for idx in df[mhz_in_specs].index:
    spec = df.loc[idx, 'Specifications']

    # Look for MHz values
    mhz_match = re.search(r'(\d+\.?\d*)\s*MHz', spec, re.IGNORECASE)

    if mhz_match and pd.isna(df.loc[idx, 'cpu_speed_ghz']):
        mhz_value = float(mhz_match.group(1))

        # Convert MHz to GHz
        ghz_value = mhz_value / 1000

        df.loc[idx, 'cpu_speed_ghz'] = ghz_value
        print(f"  Index {idx}: Converted {mhz_value} MHz to {ghz_value} GHz")

print("\n✓ Converted MHz to GHz")
```

```python
# Problem 27: Fill 0 for phones with "No Rear Camera"

print("Filling 0 for 'No Rear Camera' phones...")

# Find phones with "No Rear Camera" text
no_rear_camera = df['Specifications'].str.contains(
    'No Rear Camera', case=False, na=False)
missing_rear = df['rear_camera_mp'].isna()

needs_zero_fill = no_rear_camera & missing_rear

print(f"Phones with 'No Rear Camera': {needs_zero_fill.sum()}")

if needs_zero_fill.sum() > 0:
    print(f"Indices: {df[needs_zero_fill].index.tolist()}")

    # Fill with 0
    df.loc[needs_zero_fill, 'rear_camera_mp'] = 0.0

print(f"✓ Filled {needs_zero_fill.sum()} phones with 0 for rear_camera_mp")
```

```python
# Problem 28: Add column to indicate number of rear cameras

print("Adding column for rear camera count...")

# Create new column for rear camera count
df['rear_camera_count'] = 1  # Default single camera

# Extract camera count from specifications
camera_counts = {
    'Penta': 5,
    'Quad': 4,
    'Triple': 3,
    'Dual': 2
}

for camera_type, count in camera_counts.items():
    pattern = f'{camera_type}\\s+Rear'
    has_multi = df['Specifications'].str.contains(
        pattern, case=False, na=False, regex=True)
    df.loc[has_multi, 'rear_camera_count'] = count

# If rear_camera_mp is 0 or NaN, set count to 0
df.loc[df['rear_camera_mp'].isna() | (df['rear_camera_mp'] == 0),
       'rear_camera_count'] = 0

print("Rear camera count distribution:")
print(df['rear_camera_count'].value_counts().sort_index())

print("\n✓ Added rear_camera_count column")
```

```python
# Problem 29: Normalize OS names - remove decimal precision

print("Normalizing OS names (removing decimal precision)...")

# Count OS with decimal precision
os_with_decimal = df['operating_system'].str.contains(r'v\d+\.\d+',
                                                      case=False, na=False, regex=True)
print(f"OS values with decimal precision: {os_with_decimal.sum()}")

if os_with_decimal.sum() > 0:
    print("\nBefore normalization (sample):")
    print(df[os_with_decimal]['operating_system'].unique()[:10])

    # Remove decimal precision - keep only major version
    # Pattern: "Android v14.0" -> "Android v14"
    df['operating_system'] = df['operating_system'].str.replace(
        r'(v)(\d+)\.\d+', r'\g<1>\g<2>', regex=True
    )

    print("\nAfter normalization (sample):")
    print(df[df['operating_system'].notna()]['operating_system'].unique()[:10])

print("\n✓ Normalized OS names to integer versions only")
```

```python
# Problem 30: Extract Symbian OS from specifications

print("Extracting Symbian OS...")

# Find phones with Symbian in specs
symbian_specs = df['Specifications'].str.contains(
    'Symbian', case=False, na=False)
missing_os = df['operating_system'].isna()

needs_symbian = symbian_specs & missing_os

print(f"Phones with Symbian OS to extract: {needs_symbian.sum()}")

if needs_symbian.sum() > 0:
    print(f"Indices: {df[needs_symbian].index.tolist()}")

    for idx in df[needs_symbian].index:
        spec = df.loc[idx, 'Specifications']

        # Extract Symbian version if available
        symbian_match = re.search(
            r'Symbian(?:\s+v?(\d+))?', spec, re.IGNORECASE)

        if symbian_match:
            if symbian_match.group(1):
                os_value = f"Symbian v{symbian_match.group(1)}"
            else:
                os_value = "Symbian"

            df.loc[idx, 'operating_system'] = os_value
            print(f"  Index {idx}: Extracted {os_value}")

print(f"\n✓ Extracted {needs_symbian.sum()} Symbian OS values")
```

```python
# Problem 33: Add column for secondary front camera (Dual front camera)

print("Adding front_camera_secondary_mp column...")

# Create new column
df['front_camera_secondary_mp'] = 0.0  # Default: no secondary camera

# Find phones with dual/triple front cameras
dual_triple_front = df['Specifications'].str.contains(r'(Dual|Triple)\s+Front\s*Camera',
                                                      case=False, na=False, regex=True)

print(f"Phones with Dual/Triple Front Camera: {dual_triple_front.sum()}")

if dual_triple_front.sum() > 0:
    for idx in df[dual_triple_front].index:
        spec = df.loc[idx, 'Specifications']

        # Try to extract secondary camera value
        # Pattern: XX MP + YY MP ... Front Camera
        # Look for two MP values before "Front Camera"
        matches = re.findall(r'(\d+\.?\d*)\s*MP', spec)

        # Find the part with Front Camera
        if 'Front Camera' in spec:
            front_part = spec[spec.find('&'):] if '&' in spec else spec
            front_matches = re.findall(r'(\d+\.?\d*)\s*MP', front_part)

            # If we found multiple MP values before Front Camera
            if len(front_matches) >= 2:
                # First is primary (already in front_camera_mp)
                # Second is secondary
                secondary_mp = float(front_matches[1])
                df.loc[idx, 'front_camera_secondary_mp'] = secondary_mp

                if idx % 20 == 0:  # Show sample
                    print(
                        f"  Index {idx}: Primary={front_matches[0]} MP, Secondary={secondary_mp} MP")

print(f"\n✓ Added front_camera_secondary_mp column")
print(
    f"Phones with secondary front camera: {(df['front_camera_secondary_mp'] > 0).sum()}")
```

```python
# Problem 34: Convert resolution columns to Int64 (nullable integer)

print("Converting resolution columns to Int64...")

print(f"Before: resolution_width dtype = {df['resolution_width'].dtype}")
print(f"Before: resolution_height dtype = {df['resolution_height'].dtype}")

# Convert to Int64 (pandas nullable integer type)
df['resolution_width'] = df['resolution_width'].astype('Int64')
df['resolution_height'] = df['resolution_height'].astype('Int64')

print(f"\nAfter: resolution_width dtype = {df['resolution_width'].dtype}")
print(f"After: resolution_height dtype = {df['resolution_height'].dtype}")

print("\n✓ Converted resolution columns to Int64")
```

```python
# Problem 35: Convert card_supported to Int64

print("Converting card_supported to Int64...")

print(f"Before: card_supported dtype = {df['card_supported'].dtype}")

# Convert to Int64 (pandas nullable integer type)
df['card_supported'] = df['card_supported'].astype('Int64')

print(f"After: card_supported dtype = {df['card_supported'].dtype}")

print("\n✓ Converted card_supported to Int64")
```

```python
# Problem 37: Add columns for secondary and tertiary rear cameras

print("Adding rear_camera_secondary_mp and rear_camera_tertiary_mp columns...")

# Initialize columns
df['rear_camera_secondary_mp'] = 0.0
df['rear_camera_tertiary_mp'] = 0.0

# Find phones with multiple rear cameras
multi_rear = df['rear_camera_count'] >= 2

print(f"Phones with 2+ rear cameras: {multi_rear.sum()}")

if multi_rear.sum() > 0:
    for idx in df[multi_rear].index:
        spec = df.loc[idx, 'Specifications']

        # Find the rear camera part (before "&" or "Rear")
        if 'Rear' in spec:
            # Extract all MP values before "Rear"
            rear_part = spec.split('&')[0] if '&' in spec else spec

            # Find all MP values in rear camera part
            mp_values = re.findall(r'(\d+\.?\d*)\s*MP', rear_part)

            # Only keep values that appear before "Front"
            if 'Front' in spec:
                front_pos = spec.find('Front')
                rear_part = spec[:front_pos]
                mp_values = re.findall(r'(\d+\.?\d*)\s*MP', rear_part)

            if len(mp_values) >= 2:
                # First is primary (already in rear_camera_mp)
                # Second is secondary
                df.loc[idx, 'rear_camera_secondary_mp'] = float(mp_values[1])

                if len(mp_values) >= 3:
                    # Third is tertiary
                    df.loc[idx, 'rear_camera_tertiary_mp'] = float(
                        mp_values[2])

                if idx % 100 == 0:  # Show sample
                    print(
                        f"  Index {idx}: {mp_values[0]}/{mp_values[1] if len(mp_values) > 1 else 0}/{mp_values[2] if len(mp_values) > 2 else 0} MP")

print(f"\n✓ Added rear camera secondary/tertiary columns")
print(
    f"Phones with secondary rear camera: {(df['rear_camera_secondary_mp'] > 0).sum()}")
print(
    f"Phones with tertiary rear camera: {(df['rear_camera_tertiary_mp'] > 0).sum()}")
```

```python
# Create device_type (VERY IMPORTANT)

# keypad phones usually have very small display & no OS
df['device_type'] = 'smartphone'

df.loc[
    (df['display_size_inches'] <= 3) | 
    (df['operating_system'].isna()),
    'device_type'
] = 'feature_phone'

df['device_type'].value_counts()
```

<h1>1️⃣ Relational Database Model Theory</h1>

<p>The <strong>Relational Model</strong> was proposed by:</p>

<p>👉 Edgar F. Codd in 1970.</p>

<p>It changed the world of databases.</p>

<hr>

<h2>🔹 What is a Relational Database?</h2>

<p>A <strong>Relational Database</strong> stores data in the form of:</p>

<ul>
<li>Tables (Relations)</li>
<li>Rows (Tuples)</li>
<li>Columns (Attributes)</li>
<li>Relationships between tables</li>
</ul>

<p>Everything is mathematically based on <strong>set theory and predicate logic</strong>.</p>

<hr>

<h2>🔹 Core Concepts of Relational Model</h2>

<h3>1️⃣ Relation (Table)</h3>

<p>A relation is a table with:</p>

<ul>
<li>Unique name</li>
<li>Fixed columns</li>
<li>Rows of data</li>
</ul>

<p>Example:</p>

<table border="1">
<tr><th>id</th><th>name</th><th>age</th></tr>
<tr><td>1</td><td>Ali</td><td>21</td></tr>
<tr><td>2</td><td>Ahmed</td><td>22</td></tr>
</table>

<p>This entire structure is a <strong>relation</strong>.</p>

<hr>

<h3>2️⃣ Tuple (Row)</h3>

<p>A single record in a table.</p>

<pre>
(1, "Ali", 21)
</pre>

<hr>

<h3>3️⃣ Attribute (Column)</h3>

<p>A column in the table.</p>

<p>Example:</p>

<ul>
<li>id</li>
<li>name</li>
<li>age</li>
</ul>

<p>Each attribute has:</p>

<ul>
<li>Name</li>
<li>Data type</li>
<li>Domain (allowed values)</li>
</ul>

<hr>

<h3>4️⃣ Domain</h3>

<p>Set of allowed values for a column.</p>

<p>Example:</p>

<ul>
<li>age → only integers</li>
<li>email → string format</li>
<li>gender → {M, F}</li>
</ul>

<hr>

<h3>5️⃣ Degree</h3>

<p>Number of columns in a table.</p>

<p>If table has 3 columns → degree = 3</p>

<hr>

<h3>6️⃣ Cardinality</h3>

<p>Number of rows in a table.</p>

<p>If table has 100 rows → cardinality = 100</p>

<hr>

<h2>🔹 Properties of Relational Tables</h2>

<p>✔ Rows are unique<br>
✔ Order of rows does not matter<br>
✔ Order of columns does not matter<br>
✔ Each column has atomic values (1NF rule)<br>
✔ Each table has a primary key</p>

<hr>

<h1>2️⃣ Relational Integrity Rules</h1>

<p>Codd defined two major integrity rules:</p>

<h3>1️⃣ Entity Integrity</h3>

<ul>
<li>Primary key cannot be NULL.</li>
<li>Every row must be uniquely identifiable.</li>
</ul>

<h3>2️⃣ Referential Integrity</h3>

<ul>
<li>Foreign key must match a primary key in another table.</li>
<li>Prevents orphan records.</li>
</ul>

<p>Example:</p>

<p>Customers table<br>
Orders table referencing customer_id</p>

<h1>1️⃣ What is SQL?</h1>

<h2>📌 Definition</h2>

<p>👉 <strong>SQL (Structured Query Language)</strong> is a <strong>standard declarative language</strong> used to communicate with a relational database management system (RDBMS).</p>

<p>It is used to:</p>

<ul>
<li>Create databases</li>
<li>Create tables</li>
<li>Insert data</li>
<li>Retrieve data</li>
<li>Update data</li>
<li>Delete data</li>
<li>Manage permissions</li>
<li>Control transactions</li>
</ul>

<p>SQL works with relational databases like:</p>

<ul>
<li>MySQL</li>
<li>PostgreSQL</li>
<li>Oracle Database</li>
<li>Microsoft SQL Server</li>
</ul>

<hr>

<h1>2️⃣ Why SQL Was Created?</h1>

<p>Before SQL, databases were hierarchical or network-based.</p>

<p>The relational model was proposed by:</p>

<p>👉 Edgar F. Codd (1970)</p>

<p>Later, SQL was developed at IBM in the 1970s.</p>

<p>It became ANSI and ISO standard.</p>

<h1>3️⃣ SQL is a Declarative Language (Very Important 🔥)</h1>

<p>SQL is NOT procedural.</p>

<p>You do not say:</p>

<blockquote>
How to get data
</blockquote>

<p>You say:</p>

<blockquote>
What data you want
</blockquote>

<p>Example:</p>

<pre><code>SELECT name FROM students WHERE age &gt; 20;</code></pre>

<p>You do NOT specify:</p>

<ul>
<li>How to scan rows</li>
<li>Which index to use</li>
<li>Which algorithm to use</li>
</ul>

<p>The DBMS decides that internally.</p>

<p>That is why SQL is <strong>declarative</strong>.</p>

<h1>1️⃣ DDL – Data Definition Language</h1>

<p>DDL commands define the structure of database objects.</p>

<p>They modify <strong>schema</strong>, not data.</p>

<h2>Common DDL Commands:</h2>

<ul>
<li>CREATE</li>
<li>ALTER</li>
<li>DROP</li>
<li>TRUNCATE</li>
<li>RENAME</li>
</ul>

<hr>

<h2>🔹 CREATE</h2>

<p>Used to create:</p>

<ul>
<li>Database</li>
<li>Table</li>
<li>View</li>
<li>Index</li>
</ul>

<p>Example:</p>

<pre><code>CREATE DATABASE college;

CREATE TABLE students (
    id INT PRIMARY KEY,
    name VARCHAR(100)
);</code></pre>

<p>Internally:</p>

<ul>
<li>Allocates disk space</li>
<li>Updates system catalog</li>
<li>Creates metadata entries</li>
</ul>

<hr>

<h2>🔹 ALTER</h2>

<p>Modifies structure.</p>

<pre><code>ALTER TABLE students ADD age INT;</code></pre>

<hr>

<h2>🔹 DROP</h2>

<p>Deletes complete object permanently.</p>

<pre><code>DROP TABLE students;</code></pre>

<p>Deletes:</p>

<ul>
<li>Table structure</li>
<li>All data</li>
<li>Indexes</li>
</ul>

<hr>

<h2>🔹 TRUNCATE</h2>

<p>Deletes all rows but keeps structure.</p>

<pre><code>TRUNCATE TABLE students;</code></pre>

<p>Difference:</p>

<ul>
<li>TRUNCATE is faster</li>
<li>Cannot use WHERE</li>
<li>Usually auto-commits</li>
</ul>

<hr>

<h2>🔹 RENAME</h2>

<p>Changes name of table.</p>

<hr>

<h2>Important Theory About DDL</h2>

<ul>
<li>DDL commands are usually auto-commit.</li>
<li>Cannot be rolled back in most systems.</li>
<li>They update the <strong>data dictionary</strong>.</li>
</ul>

<hr>

<h1>2️⃣ DML – Data Manipulation Language</h1>

<p>DML deals with actual data inside tables.</p>

<h2>Commands:</h2>

<ul>
<li>INSERT</li>
<li>UPDATE</li>
<li>DELETE</li>
</ul>

<hr>

<h2>🔹 INSERT</h2>

<p>Adds new rows.</p>

<pre><code>INSERT INTO students VALUES (1, 'Ali');</code></pre>

<p>Internally:</p>

<ul>
<li>Checks constraints</li>
<li>Writes to buffer pool</li>
<li>Logs transaction</li>
</ul>

<hr>

<h2>🔹 UPDATE</h2>

<p>Modifies existing rows.</p>

<pre><code>UPDATE students SET name = 'Ahmed' WHERE id = 1;</code></pre>

<p>Without WHERE → updates all rows.</p>

<hr>

<h2>🔹 DELETE</h2>

<p>Removes specific rows.</p>

<pre><code>DELETE FROM students WHERE id = 1;</code></pre>

<p>Difference:</p>

<ul>
<li>DELETE can use WHERE</li>
<li>Logs each row removal</li>
</ul>

<hr>

<h2>Important Theory About DML</h2>

<ul>
<li>DML works inside transactions.</li>
<li>Can be rolled back.</li>
<li>Changes actual table data.</li>
</ul>

<h1>4️⃣ CREATE DATABASE</h1>

<h2>📌 Definition</h2>

<p>Creates a new logical database in the DBMS.</p>

<hr>

<h2>Basic Syntax (MySQL Example)</h2>

<pre><code>CREATE DATABASE college;</code></pre>

<p>Using IF NOT EXISTS:</p>

<pre><code>CREATE DATABASE IF NOT EXISTS college;</code></pre>

<hr>

<h2>With Character Set &amp; Collation</h2>

<pre><code>CREATE DATABASE college
CHARACTER SET utf8mb4
COLLATE utf8mb4_unicode_ci;</code></pre>

<hr>

<h2>What Happens Internally? 🔥</h2>

<p>When you run CREATE DATABASE in:</p>

<p>👉 MySQL</p>

<p>Internally:</p>

<ol>
<li>MySQL checks user privileges</li>
<li>Updates system catalog (data dictionary)</li>
<li>Creates directory inside data folder:</li>
</ol>

<pre>
/mysql/data/college/
</pre>

<ol start="4">
<li>Allocates metadata</li>
<li>Commits automatically</li>
</ol>

<p>So physically, a folder is created.</p>

<hr>

<h2>Important Theory</h2>

<ul>
<li>Database name must be unique.</li>
<li>Cannot contain special characters (except <code>_</code>).</li>
<li>Case sensitivity depends on OS.</li>
<li>It is auto-commit.</li>
<li>Cannot be rolled back.</li>
</ul>

<hr>

<h1>5️⃣ DROP DATABASE</h1>

<h2>📌 Definition</h2>

<p>Deletes entire database permanently.</p>

<hr>

<h2>Syntax</h2>

<pre><code>DROP DATABASE college;</code></pre>

<p>Safe version:</p>

<pre><code>DROP DATABASE IF EXISTS college;</code></pre>

<hr>

<h2>What Happens Internally? 🔥</h2>

<ol>
<li>Checks privileges</li>
<li>Removes database metadata from system tables</li>
<li>Deletes entire folder</li>
<li>Deletes all:</li>
</ol>

<ul>
<li>Tables</li>
<li>Views</li>
<li>Indexes</li>
<li>Stored procedures</li>
<li>Triggers</li>
</ul>

<p>⚠ No recovery unless backup exists.</p>

<hr>

<h2>Important Theory</h2>

<ul>
<li>Auto-commit</li>
<li>Irreversible</li>
<li>Very dangerous in production</li>
<li>Requires DROP privilege</li>
</ul>

<hr>

<h1>6️⃣ ALTER DATABASE</h1>

<h2>📌 Definition</h2>

<p>Modifies database properties.</p>

<p>You cannot:</p>

<ul>
<li>Rename database directly in MySQL (older versions)</li>
<li>Change everything</li>
</ul>

<p>Mostly used to:</p>

<ul>
<li>Change character set</li>
<li>Change collation</li>
</ul>

<hr>

<h2>Syntax</h2>

<pre><code>ALTER DATABASE college
CHARACTER SET = utf8mb4
COLLATE = utf8mb4_general_ci;</code></pre>

<hr>

<h2>What Happens Internally?</h2>

<ul>
<li>Updates metadata</li>
<li>Affects default character encoding for new tables</li>
<li>Does NOT change existing tables automatically</li>
</ul>

<hr>

<h1>7️⃣ USE Command (Context Switching)</h1>

<p>Technically not DDL, but extremely important.</p>

<pre><code>USE college;</code></pre>

<p>What it does:</p>

<ul>
<li>Sets active database</li>
<li>All table operations apply to this DB</li>
</ul>

<p>Internally:</p>

<ul>
<li>Changes session context</li>
<li>No physical changes</li>
</ul>

<hr>

<h1>8️⃣ Database Privileges (Advanced Concept)</h1>

<p>Database-level permissions include:</p>

<ul>
<li>CREATE</li>
<li>DROP</li>
<li>ALTER</li>
<li>GRANT OPTION</li>
</ul>

<p>Example:</p>

<pre><code>GRANT ALL PRIVILEGES ON college.* TO user1;</code></pre>

<p>This allows user to manage database objects inside that database.</p>

<hr>

<h1>9️⃣ System Catalog / Data Dictionary (Very Important 🔥)</h1>

<p>All database information is stored in system tables.</p>

<p>In MySQL, metadata stored inside:</p>

<ul>
<li>information_schema</li>
<li>mysql system database</li>
</ul>

<p>Example:</p>

<pre><code>SELECT * FROM information_schema.SCHEMATA;</code></pre>

<p>Shows all databases.</p>

<p>So CREATE DATABASE updates SCHEMATA table internally.</p>

<h1>1️⃣1️⃣ ACID and Database-Level DDL</h1>

<p>Important fact:</p>

<ul>
<li>Most DDL commands auto-commit.</li>
<li>You cannot rollback CREATE DATABASE.</li>
<li>DDL implicitly commits active transaction.</li>
</ul>

<p>Example:</p>

<pre>

CREATE DATABASE test;
ROLLBACK;
</pre>

<p>Database will still exist.</p>

<h1>1️⃣ FOREIGN KEY Constraint</h1>

<h2>Definition</h2>

<p>A <strong>FOREIGN KEY</strong> is a constraint that creates a <strong>relationship between two tables</strong>.</p>

<p>It ensures that a value in one table <strong>must exist in another table</strong>.</p>

<p>It maintains <strong>referential integrity</strong>.</p>

<h2>Basic Concept</h2>

<p>Two tables are involved:</p>

<h3>Parent Table</h3>

<p>Contains <strong>PRIMARY KEY or UNIQUE key</strong></p>

 id INT PRIMARY KEY,

<p>Contains <strong>FOREIGN KEY</strong></p>

);</code></pre>

</table>

<tr><th>id</th><th>dept_id</th></tr>

<p>Because department 5 does not exist.</p>

### <b> <center> <i>Referential Actions: </b> </center></i>
Foreign key referential actions determine **what happens to dependent rows in a child table when the referenced primary key in the parent table is updated or deleted.** These actions are specified using the ON DELETE and ON UPDATE clauses when defining a FOREIGN KEY constraint. 
1. **RESTRICT**
2. **CASCADE**
3. **SET NULL**
4. **SET DEFAULT**

#### <center>1. Restrict (Default)</center>
The delete or update operation on the parent table is rejected if any related rows exist in the child table. The check is performed before the operation begins.

```python
# cannot delete the reference table
# cursor.execute("DROP TABLE customers") 
```

#### <center>2. Cascade</center>
Deleting or updating a row in the parent table automatically deletes or updates the matching rows in the child table.

```python
cursor.execute("DROP TABLE orders")
```

```python
orders_table = """
CREATE TABLE orders(
    order_id INT PRIMARY KEY AUTO_INCREMENT,
    cid INT NOT NULL,
    order_date DATETIME DEFAULT CURRENT_TIMESTAMP,
    
    CONSTRAINT order_fk FOREIGN KEY (cid) REFERENCES customers(cid)
    ON DELETE CASCADE
    ON UPDATE CASCADE
)
"""

cursor.execute(orders_table)
```

```python
orders = """
INSERT INTO orders (order_id, cid) VALUES
(1, 2),
(2, 1);
"""
cursor.execute(orders)
```

```python
cursor.execute("Select * from orders")

for row in cursor:
    print(row)
```

<span style="font-size: 18px;"> <b> What Happens? If you delete a customer: No orphan rows allowed in referral table

```python
cursor.execute("DELETE FROM customers WHERE cid = 1;")
```

<span style="font-size: 18px;"> <b>As you see data gets deleted automatically in the orders table

```python
cursor.execute("Select * from orders")

for row in cursor:
    print(row)
```

```python
cursor.execute("UPDATE customers SET cid = 10 WHERE cid = 2;")

cursor.execute("Select * from orders")

for row in cursor:
    print(row)
```

#### <center>3. Set NULL</center>
When a row in the parent table is deleted or updated, the foreign key column(s) in the dependent child table are automatically set to NULL. This action is only possible if the foreign key column(s) are nullable.

```python
cursor.execute("DROP TABLES IF EXISTS orders")
cursor.execute("DROP TABLES IF EXISTS customers")
```

```python
customer_table = """
CREATE TABLE customers(
    cid INT PRIMARY KEY AUTO_INCREMENT,
    name VARCHAR(255) NOT NULL,
    email VARCHAR(255) NOT NULL UNIQUE
)
"""

cursor.execute(customer_table)

orders_table = """
CREATE TABLE IF NOT EXISTS orders(
    order_id INT PRIMARY KEY AUTO_INCREMENT,
    cid INT,
    order_date DATETIME DEFAULT CURRENT_TIMESTAMP,
    
    CONSTRAINT order_fk FOREIGN KEY (cid) REFERENCES customers(cid)
    ON DELETE SET NULL
    ON UPDATE SET NULL
)
"""

cursor.execute(orders_table)

customer = """
INSERT INTO customers (name, email) VALUES
('Ali Zain', 'zain@gmail.com'),
('zain', 'ali@gmail.com');
"""
cursor.execute(customer)

orders = """
INSERT INTO orders (order_id, cid) VALUES
(1, 2),
(2, 1);
"""
cursor.execute(orders)
```

```python
cursor.execute("Select * from orders")

for row in cursor:
    print(row)
```

```python
cursor.execute("DELETE FROM customers WHERE cid = 1;")

cursor.execute("Select * from orders")

for row in cursor:
    print(row)
```

## <Center>Adding Constraints

```python
cursor.execute("ALTER TABLE customers ADD COLUMN age INT NOT NULL")
```

<span style="font-size: 20px;"> <b>Add constraints so the age of the customer should be above 21

```python
cursor.execute("ALTER TABLE customers ADD CONSTRAINT customer_age_check CHECK (age > 21)")
```

## <center> Drop Constraints

<span style="font-size: 20px;"> <b>Delete the age constraint

```python
cursor.execute("ALTER TABLE customers DROP CONSTRAINT customer_age_check")
```

## <center> Modify Constraints

In MySQL, you **cannot directly "modify" an existing named constraint** (like a FOREIGN KEY, UNIQUE, or CHECK constraint) in a single step. You must instead use a **drop-and-recreate approach**, which involves dropping the constraint first and then adding it back with the new definition.<BR>
However, for NOT NULL constraints, you can use ALTER TABLE ... MODIFY to toggle them on or off. 

<span style="font-size: 20px;"> <b>Change the accepted age from 21 to 13

```python
# first add the constraint as we have delete it in above cell
cursor.execute("ALTER TABLE customers ADD CONSTRAINT customer_age_check Check (age > 21)")
```

```python
cursor.execute("SHOW CREATE TABLE customers")

result = cursor.fetchone()
print(result[1])
```

```python
cursor.execute("ALTER TABLE customers DROP CONSTRAINT customer_age_check")
cursor.execute("ALTER TABLE customers ADD CONSTRAINT customer_age_check CHECK (age > 13)")
```

```python
cursor.execute("SHOW CREATE TABLE customers")

result = cursor.fetchone()
print(result[1])
```

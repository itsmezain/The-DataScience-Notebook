For running SQL queries inside a Jupyter Notebook (`.ipynb`) with a clean, revisable workflow, the best library is **`ipython-sql`** combined with **`SQLAlchemy`**.

***

## Why `ipython-sql`?

It lets you write raw SQL directly in notebook cells using `%%sql` magic commands â€” no `cursor.execute()`, no `fetchall()`, no boilerplate. Output is rendered as a clean table automatically, perfect for revision notes.

***

## Installation

```bash
pip install ipython-sql sqlalchemy pymysql
```

***

## Setup in Notebook

```python
# Cell 1 â€” Load the extension
%load_ext sql

# Cell 2 â€” Connect to your MySQL database
%sql mysql+pymysql://root:yourpassword@localhost/smartphone
```

After this, every cell can run SQL directly.

***

## Two Ways to Write Queries

**Single-line query (`%sql`):**
```python
%sql SELECT * FROM mobiles LIMIT 5;
```

**Multi-line query (`%%sql`):**
```python
%%sql
SELECT brand_name, AVG(Price) as avg_price
FROM mobiles
GROUP BY brand_name
ORDER BY avg_price DESC
LIMIT 10;
```

Output is automatically rendered as a formatted table in the notebook.

***

## Storing Query Results into Pandas DataFrame

This is very useful for EDA after fetching data:

```python
result = %sql SELECT * FROM mobiles
df = result.DataFrame()
df.head()
```

***

## Important Things to Know

### 1. Connection String Format
```
mysql+pymysql://username:password@host/database_name
```
- `pymysql` is the driver (you can also use `mysqlconnector` â†’ `mysql+mysqlconnector://...`)
- Always specify the database at the end to avoid `USE dbname` every time

### 2. Multiple Databases
You can switch databases mid-notebook:
```python
%sql USE smartphone
```
Or reconnect to a different DB:
```python
%sql mysql+pymysql://root:pass@localhost/another_db
```

### 3. Parameterized Queries (Using Python Variables)
```python
brand = 'Samsung'
%sql SELECT * FROM mobiles WHERE brand_name = :brand
```
Use `:variable_name` syntax to inject Python variables safely.

### 4. Saving Results to CSV
```python
result = %sql SELECT * FROM mobiles
df = result.DataFrame()
df.to_csv('output.csv', index=False)
```

### 5. Displaying Row Limit
By default, `ipython-sql` only shows 10 rows. Change it:
```python
%config SqlMagic.displaylimit = 50   # show 50 rows
%config SqlMagic.displaylimit = None # show all rows
```

### 6. Auto-commit Behavior
By default, DML queries (`INSERT`, `UPDATE`, `DELETE`) may not auto-commit. Force it with:
```python
%config SqlMagic.autocommit = True
```

### 7. Feedback Toggle
Turn off the "X rows affected" message for cleaner output:
```python
%config SqlMagic.feedback = False
```

# for numerical cols
    - 8 number summary[count,min,max,mean,std,q1,q2,q3]

```python
%%sql $laptop

SELECT
    COUNT(Price) AS Count,
    MIN(Price) AS Minimum,
    MAX(Price) AS Maximum,
    AVG(Price) AS Average,
    STD(Price) AS Std_Deviation
FROM laptop;
```

```python
%%sql $laptop

-- missing values

SELECT COUNT(Price)
from laptop
WHERE Price IS NULL;
```

```python
%%sql $laptop

-- outlier using z-score

WITH stats AS (
    SELECT AVG(Price) AS mean_price, STDDEV(Price) AS sd_price
    FROM laptop
)
SELECT l.*
FROM laptop l, stats
WHERE ABS(l.Price - stats.mean_price) > 3 * stats.sd_price;
```

```python
%%sql $laptop

-- outlier using IR range
WITH ordered AS (
    SELECT
        Price,
        ROW_NUMBER() OVER (ORDER BY Price) AS row_num,
        COUNT(*) OVER () AS total_rows
    FROM laptop
),
quartiles AS (
    SELECT
        MAX(CASE WHEN row_num = FLOOR(0.25 * total_rows) THEN Price END) AS Q1,
        MAX(CASE WHEN row_num = FLOOR(0.75 * total_rows) THEN Price END) AS Q3
    FROM ordered
),
bounds AS (
    SELECT
        Q1,
        Q3,
        (Q3 - Q1) AS IQR,
        (Q1 - 1.5 * (Q3 - Q1)) AS lower_bound,
        (Q3 + 1.5 * (Q3 - Q1)) AS upper_bound
    FROM quartiles
)
SELECT l.*
FROM laptop l, bounds b
WHERE l.Price < b.lower_bound OR l.Price > b.upper_bound
ORDER BY PRICE;
```

```python
%%sql $laptop

SELECT t.buckets, REPEAT('*', COUNT(*)/5) FROM (SELECT 
    Price,
    CASE
        WHEN price BETWEEN 0 AND 25000 THEN '0-25K'
        WHEN price BETWEEN 25001 AND 50000 THEN '25K-50K'
        WHEN price BETWEEN 50001 AND 75000 THEN '50K-75K'
        WHEN price BETWEEN 75001 AND 100000 THEN '75K-100K'
        ELSE '>100K'
    END as 'buckets'
FROM laptop) t
GROUP BY t.buckets
```

## categorical-categorical

```python
%%sql $laptop

SELECT 
    TypeName,
    SUM(CASE WHEN TouchScreen = 0 THEN 1 ELSE 0 END) AS No_Touchscreen,
    SUM(CASE WHEN TouchScreen = 1 THEN 1 ELSE 0 END) AS Yes_Touchscreen,
    COUNT(*) AS Row_Total
FROM laptop
GROUP BY TypeName
ORDER BY TypeName
```

```python
%%sql $laptop

-- Add grand total in the above table 
SELECT 
    TypeName,
    SUM(CASE WHEN TouchScreen = 0 THEN 1 ELSE 0 END) AS No_Touchscreen,
    SUM(CASE WHEN TouchScreen = 1 THEN 1 ELSE 0 END) AS Yes_Touchscreen,
    COUNT(*) AS Row_Total
FROM laptop
GROUP BY TypeName

UNION ALL

SELECT
    'TOTAL',
    SUM(CASE WHEN Touchscreen = 0 THEN 1 ELSE 0 END),
    SUM(CASE WHEN Touchscreen = 1 THEN 1 ELSE 0 END),
    COUNT(*)
FROM laptop
```

```python
%%sql $laptop

SELECT 
    Company,
    SUM(CASE WHEN TouchScreen = 1 THEN 1 ELSE 0 END) AS Yes_Touchscreen,
    SUM(CASE WHEN TouchScreen = 0 THEN 1 ELSE 0 END) AS No_Touchscreen,
    COUNT(*) AS Row_Total
FROM laptop
GROUP BY Company
ORDER BY Company
```

## ðŸ—ºï¸ 4ï¸âƒ£ Choropleth Map (Colored Regions)

If you have **GeoJSON** data (India district/state boundaries), you can fill regions by color (e.g., literacy rate or population).

Hereâ€™s how:

### Step 1: Load the GeoJSON (downloadable from GitHub or Kaggle)

```python
india_geojson = 'https://raw.githubusercontent.com/geohacker/india/master/state/india_telengana.geojson'
```

*(Replace with your actual India GeoJSON file â€” there are many available free online)*

### Step 2: Merge and plot

```python
fig = px.choropleth(
    df,
    geojson=india_geojson,
    featureidkey="properties.ST_NM",   # key in geojson
    locations="State name",             # key in your dataset
    color="Literacy",                   # variable to color by
    hover_name="State name",
    color_continuous_scale="Viridis",
    title="Literacy Rate Across Indian States"
)
fig.update_geos(fitbounds="locations", visible=False)
fig.show()
```

âœ… **Result:** A color-shaded India map showing literacy rate by state.

## ðŸŒ 1ï¸âƒ£ What is `scatter_mapbox()`?

Itâ€™s a Plotly Express function that plots **points on an interactive Mapbox map** (like Google Maps).

ðŸ‘‰ Each point is defined by its **latitude** and **longitude** columns.

---

## âš™ï¸ 2ï¸âƒ£ Basic Syntax

```python
import plotly.express as px
import pandas as pd

# Sample dataframe
df = pd.DataFrame({
    'District': ['Agra', 'Lucknow', 'Varanasi', 'Kanpur'],
    'State': ['Uttar Pradesh'] * 4,
    'Population': [4380793, 4589838, 3676841, 4572951],
    'Literacy': [62.56, 74.5, 77.8, 80.1],
    'lat': [27.18, 26.85, 25.32, 26.45],
    'lon': [78.02, 80.95, 83.01, 80.35]
})

# Create the map
fig = px.scatter_mapbox(
    df,
    lat='lat',
    lon='lon',
    size='Population',
    color='Literacy',
    hover_name='District',
    hover_data=['State', 'Population', 'Literacy'],
    color_continuous_scale='Viridis',
    size_max=35,
    zoom=4,
    height=600,
    title='India District Population & Literacy Rate'
)

# Use OpenStreetMap (free and works without API key)
fig.update_layout(mapbox_style='open-street-map')
fig.show()
```

âœ… This gives you a **zoomable, pannable India map** with bubbles sized by population and colored by literacy.

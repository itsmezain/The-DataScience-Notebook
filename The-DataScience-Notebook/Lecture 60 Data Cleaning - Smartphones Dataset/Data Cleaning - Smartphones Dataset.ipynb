```python
import re

def extract_processor_comprehensive(spec_text):
    """
    Comprehensive processor extraction searching entire Specifications text.
    Uses advanced regex patterns to identify processor brands/models anywhere in the spec.
    """
    if pd.isna(spec_text):
        return None
    
    spec_text = str(spec_text)
    
    # Comprehensive processor patterns - ordered by specificity (most specific first)
    patterns = [
        # Google Tensor
        (r'Tensor\s+G\d+', lambda m: m.group(0)),
        
        # Qualcomm Snapdragon (all variants)
        (r'Snapdragon\s+\d+\s*(?:Elite|Plus|Ultra)', lambda m: m.group(0)),
        (r'Snapdragon\s+\d+[a-z]\s+Gen\s*\d+', lambda m: m.group(0)),
        (r'Snapdragon\s+\d+\s+Gen\s*\d+', lambda m: m.group(0)),
        (r'Snapdragon\s+\d+', lambda m: m.group(0)),
        
        # MediaTek Dimensity
        (r'Dimensity\s+\d+(?:\s+(?:Plus|Ultra|Energy|Ultimate))?', lambda m: m.group(0)),
        
        # MediaTek Helio  
        (r'Helio\s+[A-Z]\d+(?:\s+(?:Plus|Ultra))?', lambda m: m.group(0)),
        
        # Samsung Exynos
        (r'Exynos\s+\d+', lambda m: m.group(0)),
        
        # Unisoc (all variants - very specific patterns first)
        (r'Unisoc\s+\d+\s+FF[\-\s]*[A-Z]', lambda m: m.group(0).replace(' ', ' ').strip()),  # "8910 FF-S"
        (r'Unisoc\s+SC\d+[A-Z]+', lambda m: m.group(0)),  # "SC6531E"
        (r'Unisoc\s+T\d+[A-Z]?', lambda m: m.group(0)),  # "T7250"
        (r'Unisoc\s+[A-Z]?\d+[\w\-]*(?=\s|,|$)', lambda m: m.group(0)),  # Other Unisoc variants
        (r'\bUnisoc\b(?=\s*,|\s+\d|\s*$)', lambda m: 'Unisoc'),  # Just "Unisoc" alone
        
        # Apple (various formats)
        (r'Bionic\s+A\d+(?:\s+Pro)?', lambda m: m.group(0)),
        (r'Apple\s+A\d+(?:\s+Pro)?', lambda m: m.group(0)),
        (r'A\d+\s+Bionic', lambda m: m.group(0)),
        
        # Huawei Kirin
        (r'Kirin\s+\d+', lambda m: m.group(0)),
        
        # General MediaTek
        (r'MediaTek\s+\w+\d+', lambda m: m.group(0)),
    ]
    
    # Try each pattern
    for pattern, formatter in patterns:
        match = re.search(pattern, spec_text, re.IGNORECASE)
        if match:
            processor = formatter(match)
            # Normalize spacing
            processor = re.sub(r'\s+', ' ', processor).strip()
            return processor
    
    return None


print("="*80)
print("STEP 1: Filling NULL processor names from Specifications")
print("="*80)

null_count_before = df['processor_name'].isna().sum()
filled_from_spec = 0

for idx in df[df['processor_name'].isna()].index:
    spec = df.loc[idx, 'Specifications']
    extracted = extract_processor_comprehensive(spec)
    
    if extracted:
        print(f"   {df.loc[idx, 'Product_Name'][:60]:<60}  {extracted}")
        df.loc[idx, 'processor_name'] = extracted
        filled_from_spec += 1

print(f"\n Filled {filled_from_spec} null values")
print(f" Remaining nulls: {df['processor_name'].isna().sum()} (likely basic phones without processor info)")


print("\n" + "="*80)
print("STEP 2: Fixing WRONG processor names (Octa Core, No Wifi, etc.)")
print("="*80)

wrong_values = ['Octa Core', 'No Wifi', 'Quad Core', 'Single Core', 'Dual Core', 'Hexa Core']
fixed_count = 0

for wrong_val in wrong_values:
    wrong_indices = df[df['processor_name'] == wrong_val].index
    
    if len(wrong_indices) > 0:
        print(f"\nFixing '{wrong_val}' ({len(wrong_indices)} cases):")
        
        for idx in wrong_indices:
            spec = df.loc[idx, 'Specifications']
            extracted = extract_processor_comprehensive(spec)
            
            if extracted:
                print(f"   {df.loc[idx, 'Product_Name'][:50]:<50} '{wrong_val}'  '{extracted}'")
                df.loc[idx, 'processor_name'] = extracted
                fixed_count += 1
            else:
                print(f"   {df.loc[idx, 'Product_Name'][:50]:<50} No valid processor found  set to NULL")
                df.loc[idx, 'processor_name'] = None


print("\n" + "="*80)
print("FINAL SUMMARY")
print("="*80)
print(f"Initial null count: {null_count_before}")
print(f"Filled from Specifications: {filled_from_spec}")
print(f"Fixed wrong values: {fixed_count}")
print(f"Final null count: {df['processor_name'].isna().sum()}")

print(f"\n Top 15 processor names after cleaning:")
print(df['processor_name'].value_counts().head(15))
```

<span style="font-size: 18px;"> <b>Extracting Processor Brand 

```python
import re


def extract_processor_brand(processor_name):
    """
    Extract processor brand from processor name.
    Handles various formats including prefixes like 'Google', 'Qualcomm', etc.
    """
    if pd.isna(processor_name):
        return None
    
    processor_name = str(processor_name).strip()
    
    # Brand patterns - ordered by specificity
    brand_patterns = [
        # Google Tensor (with or without "Google" prefix)
        (r'Google\s+Tensor|^Tensor', 'Tensor'),
        
        # Qualcomm Snapdragon (with or without "Qualcomm" prefix)
        (r'Qualcomm\s+Snapdragon|^Snapdragon', 'Snapdragon'),
        
        # MediaTek variants
        (r'^Dimensity', 'Dimensity'),
        (r'^Helio', 'Helio'),
        (r'^MTK|^MediaTek', 'MediaTek'),
        
        # Samsung
        (r'^Exynos', 'Exynos'),
        
        # Unisoc (including old Spreadtrum and Tiger series)
        (r'^Unisoc|^Spreadtrum|^Tiger', 'Unisoc'),
        
        # Apple (various formats)
        (r'^Bionic|^Apple|^A\d+\s+Bionic', 'Apple'),
        
        # Others
        (r'^Kirin', 'Kirin'),
        (r'^AMD', 'AMD'),
    ]
    
    for pattern, brand in brand_patterns:
        if re.match(pattern, processor_name, re.IGNORECASE):
            return brand
    
    # If no pattern matches and it's a wrong value, return None
    if processor_name.lower() in ['no wifi', 'octa core', 'quad core', 'single core', 'dual core', 'hexa core']:
        return None
    
    return None

# Create processor_brand column
df['processor_brand'] = df['processor_name'].apply(extract_processor_brand)

# Get position of processor_name and insert processor_brand right after it
proc_name_idx = df.columns.get_loc('processor_name')
cols = df.columns.tolist()
cols.remove('processor_brand')
cols.insert(proc_name_idx + 1, 'processor_brand')
df = df[cols]

print(" Created processor_brand column next to processor_name")
print(f"\nColumn order: {df.columns[proc_name_idx:proc_name_idx+3].tolist()}")

print("\n" + "="*80)
print("Processor brand distribution:")
print("="*80)
print(df['processor_brand'].value_counts())

print(f"\n Null processor_brand count: {df['processor_brand'].isna().sum()}")

```

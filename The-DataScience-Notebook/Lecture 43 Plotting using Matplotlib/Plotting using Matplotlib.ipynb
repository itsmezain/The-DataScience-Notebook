**Question :** Why to use ***pyplot.plot()*** for plotting a scatter plot instead of ***pyplot.scatter()*** ?  
**Answer :** ***plot()*** is faster than ***scatter()***, also plot() function plots every point uniformly but scatter plot every plot individually.
>So, by plotting each point individually you can give your plot more depth by adding extra layers of information 

Ahh, excellent 👌 This is one of those subtle but important matplotlib details that often confuses people. Let’s carefully break down the **difference between `plt.plot(x, y, '.')` and `plt.scatter(x, y)`**.

---

### 🔹 1. `plt.plot(x, y, '.')`

* **Primary purpose**: for **line plots** (connects points in order).
* If you pass `'.'` as the format string → it only plots **points (dots)** instead of lines.
* Example:

  ```python
  plt.plot(x, y, '.')
  ```

  👉 Draws points at (x,y) locations.
* Marker size is controlled by **`markersize`**:

  ```python
  plt.plot(x, y, '.', markersize=10)
  ```
* Limitations:

  * All markers must be the **same size**.
  * Cannot easily map color/size to another variable (no colormap support).

---

### 🔹 2. `plt.scatter(x, y)`

* **Primary purpose**: for **scatter plots** (statistical point visualization).
* Much more **flexible**:

  * You can vary **marker size** per point (`s=`).
  * You can vary **marker color** per point (`c=`).
  * You can use **colormaps**.
  * Supports transparency (`alpha`).
* Example:

  ```python
  plt.scatter(x, y, s=50, c='red', marker='+')
  ```
* Example with variable sizes and colors:

  ```python
  plt.scatter(x, y, s=sizes, c=colors, cmap='viridis', alpha=0.7)
  ```

---

### 🔹 Key Differences in One Table

| Feature                       | `plt.plot(x,y,'.')`                   | `plt.scatter(x,y)`                |
| ----------------------------- | ------------------------------------- | --------------------------------- |
| Main purpose                  | Line plotting, dots as a special case | Scatter visualization             |
| Marker size control           | `markersize` (all same size)          | `s` (can vary per point)          |
| Marker color control          | `color` (all same color)              | `c` (single color or array)       |
| Colormap support              | ❌ No                                  | ✅ Yes (with `c` + `cmap`)         |
| Per-point customization       | ❌ No                                  | ✅ Yes (sizes, colors, alpha, etc) |
| Performance (very large data) | ⚡ Slightly faster for uniform plots   | ⚡ More overhead, but flexible     |

---

### 🔹 Visual Example

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.rand(50)
y = np.random.rand(50)
sizes = np.linspace(10, 500, 50)   # varying sizes
colors = np.linspace(0, 1, 50)     # varying colors

plt.figure(figsize=(12,5))

# Using plot
plt.subplot(1,2,1)
plt.plot(x, y, '.', markersize=10, color="red")
plt.title("plt.plot(x, y, '.')")

# Using scatter
plt.subplot(1,2,2)
plt.scatter(x, y, s=sizes, c=colors, cmap='viridis', alpha=0.7)
plt.title("plt.scatter(x, y)")
plt.colorbar(label="Color scale")

plt.show()
```

👉 Left: simple red dots (`plt.plot`)
👉 Right: fancy scatter with varying size + color (`plt.scatter`)

---

✅ **In short:**

* Use **`plt.plot(x, y, '.')`** if you just want quick, uniform dots.
* Use **`plt.scatter(x, y)`** if you need flexible scatter plots with variable color/size/alpha.

---



Great question 👍 — this goes into how **Matplotlib handles rendering under the hood**.

Both `plt.plot` and `plt.scatter` can draw scatter-like plots, but their **implementations differ**:

---

### 🔹 1. `plt.plot(x, y, '.')`

* Internally, `plt.plot` creates a **single `Line2D` object**.
* Even when you pass `'.'` (dot marker), all points are treated as a *single line object with markers*.
* That means **one draw call to the renderer**.
* Efficient because:

  * All markers share the same size, color, and style.
  * Vectorized drawing → GPU/CPU handles them in bulk.
* For millions of points, this reduces overhead.

---

### 🔹 2. `plt.scatter(x, y)`

* Internally, `plt.scatter` creates a **`PathCollection`**.
* It is much more flexible:

  * Each point can have its own color (`c=`), size (`s=`), alpha, colormap, etc.
  * Handles mapping of values → colormap → RGBA.
* But this flexibility comes at a cost:

  * More preprocessing (scaling, color normalization).
  * Higher memory footprint.
  * More draw calls compared to `plt.plot`.
* Thus, **slower for millions of points** unless you really need per-point styling.

---

### ⚡ Why `plt.plot` can be faster

* Fewer objects to manage (1 `Line2D` vs 1 `PathCollection` with metadata for each point).
* Minimal preprocessing — just draw the markers.
* Renderer (Agg, Qt, etc.) works faster on uniform points.

---

### ⚖️ When to use which

* Use **`plt.plot(x, y, '.')`** → when you want fast scatter plotting of very large datasets where all points look the same.
* Use **`plt.scatter(x, y)`** → when you need per-point customization (sizes, colormap, categories, etc.).

---

👉 A rule of thumb:

* Millions of points → prefer `plt.plot` if uniform style.
* Fewer points / need customization → `plt.scatter`.

#### **2. Grouped Bar Chart**

````python
# Step 1: Get the x locations
x = np.arange(len(batsman_season['batsman']))  # 0,1,2,3,... for each batsman
width = 0.25   # width of each bar

# Step 2: Plot bars with shifting
plt.bar(x - width, batsman_season['2015'], width=width, color='red', label='2015')
plt.bar(x,         batsman_season['2016'], width=width, color='blue', label='2016')
plt.bar(x + width, batsman_season['2017'], width=width, color='green', label='2017')

# Step 3: Formatting
plt.xticks(x, batsman_season['batsman'], rotation=45)  # batsman names on X-axis
plt.xlabel("Batsman")
plt.ylabel("Runs Scored")
plt.title("Runs scored in IPL (2015-2017)")
plt.legend()
plt.tight_layout()
plt.show()
````

Aha 👍 you’re running into a **classic grouped bar chart problem**.
The code you wrote overlays the bars for 2015, 2016, and 2017 **on top of each other** (so only the last year’s bars are visible).

👉 What you want is to **plot them side by side** for each batsman. This is called a **grouped bar chart**. Let me walk you through step by step:

---

### Step 1. Basic Idea 🧠

* For each batsman, you need **3 bars** (2015, 2016, 2017).
* To avoid overlap, you **shift** the bars slightly left or right.
* This is done using the parameter `width` in `plt.bar`.

---

### Step 2. Syntax Explained

```python
plt.bar(x, height, width, label, color)
```

* `x` → positions on the X-axis (can be `range(len(batsmen))`)
* `height` → the actual values (runs scored)
* `width` → how thick the bar is (smaller width = space for multiple bars side by side)
* `label` → text for the legend
* `color` → bar color

---

### Step 3. Full Example

```python
import numpy as np
import matplotlib.pyplot as plt

# Suppose batsman_season looks like this:
# batsman | 2015 | 2016 | 2017
# Virat   | 640  | 700  | 800
# Rohit   | 500  | 600  | 750

# Step 1: Get the x locations
x = np.arange(len(batsman_season['batsman']))  # 0,1,2,3,... for each batsman
width = 0.25   # width of each bar

# Step 2: Plot bars with shifting
plt.bar(x - width, batsman_season['2015'], width=width, color='red', label='2015')
plt.bar(x,         batsman_season['2016'], width=width, color='blue', label='2016')
plt.bar(x + width, batsman_season['2017'], width=width, color='green', label='2017')

# Step 3: Formatting
plt.xticks(x, batsman_season['batsman'], rotation=45)  # batsman names on X-axis
plt.xlabel("Batsman")
plt.ylabel("Runs Scored")
plt.title("Runs scored in IPL (2015-2017)")
plt.legend()
plt.tight_layout()
plt.show()
```

---

### 🔎 What’s happening here:

* `x = np.arange(len(batsman_season['batsman']))` gives **index positions** of batsmen.
* We set `width = 0.25`, meaning each bar is narrow enough to fit 3 bars in a group.
* We shift the bars:

  * 2015 → `x - width` (left)
  * 2016 → `x` (center)
  * 2017 → `x + width` (right)

Now, all 3 years appear **side by side** for each batsman 🎉.

---

#### **3. Stacked Bar Chart**

1. The visualization in the above section has an alternative visualization known as ***Stacked Bar Chart***, where ***bars are stacked on each other***.  
2. Likewise Grouped Bar Chart, Matplotlib does not have easy method for this, instead, we have to ***play smart to make it happen*** as done in the case of Grouped Bar Chart.  
3. For Stacked Bar Chart, we will be playing smart with the help of ***bottom*** parameter of the ***bar()*** method which ***allows us shift the base of bars in Y-axis***.

````python
plt.bar(batsman_season['batsman'], batsman_season['2015'], bottom=0, color = 'orange', label = '2015')

plt.bar(batsman_season['batsman'], batsman_season['2016'], bottom= batsman_season['2015'], color = 'red', label = '2016')

plt.bar(batsman_season['batsman'], batsman_season['2017'], bottom= batsman_season['2015'] + batsman_season['2016'], color = 'green',  label = '2017')

plt.xticks(rotation=60)
plt.xlabel('Batsman')
plt.ylabel('Run scored')
plt.legend()
plt.grid(axis = 'y', alpha=0.8, linestyle='--', color='gray')
plt.show()
````

>Can we add number labels in the bar, so the charts are more readable.

````python
fig, ax = plt.subplots(figsize=(10,6))

b1 = ax.bar(batsman_season['batsman'], batsman_season['2015'], 
            bottom=0, color='orange', label='2015')

b2 = ax.bar(batsman_season['batsman'], batsman_season['2016'], 
            bottom=batsman_season['2015'], color='red', label='2016')

b3 = ax.bar(batsman_season['batsman'], batsman_season['2017'], 
            bottom=batsman_season['2015'] + batsman_season['2016'], 
            color='green', label='2017')

# Adding labels to each segment
ax.bar_label(b1, label_type='edge', color='orange', fontsize=10)
ax.bar_label(b2, label_type='edge', color='red', fontsize=10)
ax.bar_label(b3, label_type='edge', color='green', fontsize=10)

ax.set_xticklabels(batsman_season['batsman'], rotation=60)
ax.set_xlabel('Batsman')
ax.set_ylabel('Runs Scored')
ax.set_title('Runs scored in IPL (2015–2017)')
ax.legend()
ax.grid(axis='y', linestyle='--', alpha=0.5) 
plt.show()
````

````python
gayle = pd.read_csv('datasets\\gayle-175.csv')

gayle # gayle ne 175 runs banaya huva match ka runs
````

````python
plt.pie(gayle['batsman_runs'], labels=gayle['batsman'])

plt.show()
````

**Problem :** All slices are looking almost same. How to differentiate or analyse?  
**Solution :** Use autopct to display percentage of each slice.

````python
plt.pie(
    gayle['batsman_runs'],
    labels=None,   # don’t show labels on pie
    autopct='%1.1f%%',  # show % inside slices
)

plt.legend(gayle['batsman'], bbox_to_anchor=(1.05, 1), loc='upper left')  # labels outside
plt.show()
````

Ummm, the colors are not cool. Can we do something?  
The answer is ***Yes***.

````python
plt.pie(
    gayle['batsman_runs'],
    labels=gayle['batsman'],
    autopct='%1.1f%%',
    colors = ['blue', 'green', 'red', 'yellow', 'green', 'orange']
)

plt.legend(gayle['batsman'], bbox_to_anchor=(1.05, 1), loc='best')  # labels outside
plt.show()
````

````python
from matplotlib import colormaps
list(colormaps)
````

````python
cmap = plt.get_cmap('Greens')

plt.pie(
    x = gayle['batsman_runs'],
    labels = gayle['batsman'],
    colors = [
        cmap(i) for i in np.linspace(0.3, 3, len(gayle['batsman_runs']))
    ]
)
plt.show()
````

Let's say I want to emphasize on the slice representing highest marks. How can I do that?  
You may give a try to ***explode*** parameter of the ***pie()*** function.

````python
plt.pie(
    gayle['batsman_runs'],
    labels=gayle['batsman'],
    autopct='%1.1f%%',
    colors = ['blue', 'green', 'red', 'yellow', 'green', 'orange'],
    explode = [0.3,0,0,0,0,0]
)

plt.show()
````

>Shadows

````python
plt.pie(
    gayle['batsman_runs'],
    labels=gayle['batsman'],
    autopct='%1.1f%%',
    colors = ['blue', 'green', 'red', 'yellow', 'green', 'orange'],
    explode = [0.3,0,0,0,0,0],
    shadow = True
)

plt.show()
````

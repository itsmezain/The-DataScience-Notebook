# Pandas `dropna()`: A Complete Guide

## 1. Introduction

In real-world datasets, missing data is common. Pandas provides the `dropna()` method to **remove missing (NaN) values** from a DataFrame or Series. This is a key data cleaning step that can help improve the quality of your data before analysis or modeling.

- **Purpose:**  
  `dropna()` is used to drop rows or columns that contain missing values.
  
- **When to Use:**  
  - When you want to exclude incomplete data.
  - When missing data is not significant or can be safely ignored.
  - As part of your data cleaning pipeline.

---

## 2. Syntax and Parameters

### 2.1 Basic Syntax

For a DataFrame:
```python
DataFrame.dropna(axis=0, how='any', thresh=None, subset=None, inplace=False)
```
For a Series:
```python
Series.dropna(inplace=False)
```

### 2.2 Parameters Explained

- **`axis`**  
  - `axis=0` (default): Drop rows with missing values.
  - `axis=1`: Drop columns with missing values.

- **`how`**  
  - `'any'` (default): Drop the row/column if **any** value is NaN.
  - `'all'`: Drop the row/column only if **all** values are NaN.

- **`thresh`**  
  - Requires that a row/column has at least a certain number of **non-missing values** to be kept.
  - Example: `thresh=2` means the row/column must have at least 2 non-NaN values.

- **`subset`**  
  - Specify a list of columns or rows to consider for the missing value check.
  - Only these specified columns/rows are used to decide whether to drop the row/column.

- **`inplace`**  
  - If `True`, modify the DataFrame/Series in place (no copy is returned).
  - If `False` (default), return a new DataFrame/Series with the missing values dropped.

---

## 3. Detailed Examples

### 3.1 Dropping Rows in a DataFrame

Consider a DataFrame with missing values:
```python
import pandas as pd
import numpy as np

data = {
    'A': [1, 2, np.nan, 4],
    'B': [np.nan, 2, 3, 4],
    'C': [1, np.nan, np.nan, 4]
}
df = pd.DataFrame(data)
print("Original DataFrame:\n", df)
```
**Output:**
```
     A    B    C
0  1.0  NaN  1.0
1  2.0  2.0  NaN
2  NaN  3.0  NaN
3  4.0  4.0  4.0
```

#### **Drop Rows Where Any Value is NaN (Default)**
```python
df_drop_any = df.dropna()  
print("Drop rows with any NaN:\n", df_drop_any)
```
**Output:**
```
     A    B    C
3  4.0  4.0  4.0
```
- Only row 3 has no missing values.

#### **Drop Rows Where All Values are NaN**
```python
df_drop_all = df.dropna(how='all')
print("Drop rows with all NaN:\n", df_drop_all)
```
**Output:**  
- In this example, no row has all values missing, so the DataFrame remains unchanged.

#### **Drop Rows with a Minimum Number of Non-NaN Values**
```python
df_thresh = df.dropna(thresh=2)
print("Drop rows with fewer than 2 non-NaN values:\n", df_thresh)
```
**Output:**
```
     A    B    C
0  1.0  NaN  1.0
1  2.0  2.0  NaN
3  4.0  4.0  4.0
```
- Row 2 is dropped because it has only one non-NaN value.

### 3.2 Dropping Columns in a DataFrame

#### **Drop Columns with Any NaN Value**
```python
df_drop_cols = df.dropna(axis=1)
print("Drop columns with any NaN:\n", df_drop_cols)
```
**Output:**
```
     C
0  1.0
1  NaN
2  NaN
3  4.0
```
- Only column C is fully retained (although it has some NaNs, but here the column is kept because at least one row might be fully non-NaN if using `how='all'` might be different).

#### **Drop Columns with All NaN Values**
```python
df_drop_all_cols = df.dropna(axis=1, how='all')
print("Drop columns with all NaN:\n", df_drop_all_cols)
```
**Output:**
- Since no column has all values missing in this example, the DataFrame remains unchanged.

#### **Using `subset` Parameter for Row-Based Dropping**
```python
# Only consider columns A and B for the missing value check.
df_subset = df.dropna(subset=['A', 'B'])
print("Drop rows where A or B is NaN:\n", df_subset)
```
**Output:**
```
     A    B    C
1  2.0  2.0  NaN
3  4.0  4.0  4.0
```
- Rows 0 and 2 are dropped because they have missing values in either A or B.

### 3.3 Dropping Missing Values in a Series

For a Series:
```python
s = pd.Series([1, np.nan, 3, np.nan, 5])
print("Original Series:\n", s)

s_drop = s.dropna()
print("Series after dropna():\n", s_drop)
```
**Output:**
```
Original Series:
 0    1.0
1    NaN
2    3.0
3    NaN
4    5.0
dtype: float64

Series after dropna():
 0    1.0
2    3.0
4    5.0
dtype: float64
```

## 4. Use Cases

### **Cleaning Data for Analysis**
- Removing rows or columns with missing values to ensure that calculations (mean, sum, etc.) are not skewed.
  
### **Preprocessing for Machine Learning**
- Dropping missing values before training models to avoid errors during model fitting.

### **Data Summarization**
- Sometimes, you want to quickly summarize a dataset by eliminating incomplete data.

---

## 5. Best Practices and Considerations

- **Evaluate Missing Data:**  
  Before dropping, check how much data is missing using methods like `df.isna().sum()`.

- **Thresholds (`thresh` Parameter):**  
  Instead of dropping rows or columns that have any missing value, consider setting a threshold (`thresh`) to drop only those with too many missing values.

- **Subsetting Columns (`subset` Parameter):**  
  When you care about missing values in only specific columns, use the `subset` parameter.

- **In-Place vs. Copy:**  
  Decide whether you want to modify the original DataFrame/Series in place using `inplace=True` or create a new object.

- **Understand the Impact:**  
  Dropping missing values can reduce your dataset's size. Make sure that this does not remove too much important data.

---

## 6. Summary Table

| **Parameter** | **Description** | **Example** |
|---------------|-----------------|-------------|
| `axis` | 0 (rows) or 1 (columns) | `df.dropna(axis=0)` drops rows with NaNs |
| `how` | 'any' (default) drops row/column if any NaN exists; 'all' drops if all values are NaN | `df.dropna(how='all')` |
| `thresh` | Require at least this many non-NaN values to keep row/column | `df.dropna(thresh=2)` |
| `subset` | Specify a list of columns/rows to check for NaNs | `df.dropna(subset=['A', 'B'])` |
| `inplace` | Modify the original object if True | `df.dropna(inplace=True)` |

---

## 7. Final Thoughts
- **`dropna()`** is essential for data cleaning.
- It allows for flexible control over which missing values to drop.
- Ensure you understand the impact on your dataset before dropping data.
- Use the parameters (`axis`, `how`, `thresh`, `subset`, `inplace`) effectively to tailor your data cleaning process.

By mastering `dropna()`, you ensure that your datasets are clean and ready for analysis, modeling, or visualization. This detailed guide should now cover everything you need to know about handling missing values with `dropna()` in Pandas.

# **Pandas `fillna()`: A Complete and Detailed Guide**

Handling missing data is a crucial step in data preprocessing. Pandas provides the `fillna()` method to replace missing values (`NaN`) with a specified value or a computed value such as the mean, median, or forward/backward fill. This guide provides **detailed** and **comprehensive** notes on `fillna()` with explanations, parameters, examples, and best practices.

---

## **1. Introduction to `fillna()`**
### **What is `fillna()`?**
- `fillna()` is a method in Pandas used to **replace missing values (NaN)** in a DataFrame or Series.
- Missing values can be replaced with:
  - A **fixed value**
  - A **computed statistic** (mean, median, mode)
  - **Forward fill (propagating previous value)**
  - **Backward fill (propagating next value)**
  - A **custom function**
- `fillna()` is a more flexible alternative to simply dropping missing values (`dropna()`).

### **When to Use `fillna()`?**
- When missing values **should not be removed** but rather replaced with meaningful values.
- When performing **statistical analysis or machine learning**, where missing values can affect results.
- When preparing data for **visualization**, as missing values can disrupt plots.

---

## **2. Syntax and Parameters**
### **Basic Syntax**
For a DataFrame:
```python
DataFrame.fillna(value=None, method=None, axis=None, inplace=False, limit=None)
```
For a Series:
```python
Series.fillna(value=None, method=None, inplace=False, limit=None)
```

### **Parameters Explained**
| Parameter  | Description |
|------------|-------------|
| `value`    | A scalar, dictionary, Series, or DataFrame value used to replace NaNs. |
| `method`   | The method for filling missing values. Options: `'ffill'` (forward fill) or `'bfill'` (backward fill). |
| `axis`     | Determines whether filling is done along rows (`axis=0`, default) or columns (`axis=1`). |
| `inplace`  | If `True`, modifies the original object instead of returning a new one. |
| `limit`    | Limits the number of NaNs filled. Useful when forward/backward filling. |

---

## **3. Examples and Use Cases**

### **3.1 Replacing NaN with a Fixed Value**
```python
import pandas as pd
import numpy as np

# Creating a DataFrame with missing values
data = {
    'A': [1, 2, np.nan, 4],
    'B': [np.nan, 2, 3, 4],
    'C': [1, np.nan, np.nan, 4]
}
df = pd.DataFrame(data)
print("Original DataFrame:\n", df)

# Replace NaN with 0
df_filled = df.fillna(0)
print("Filled NaN with 0:\n", df_filled)
```
**Output:**
```

     A    B    C

1  2.0  2.0  0.0
2  0.0  3.0  0.0
3  4.0  4.0  4.0
```
✔ **Used when missing values should be replaced with a constant.**

```
✔ **Useful for numerical datasets where missing values can be replaced with statistical measures.**

```
✔ **Better for skewed data than mean.**

### **3.4 Filling with Forward Fill (`ffill`)**
```python
df_ffill = df.fillna(method='ffill')

```
✔ **Useful for time-series data where missing values should be replaced with the previous value.**

---

### **3.5 Filling with Backward Fill (`bfill`)**
```python
df_bfill = df.fillna(method='bfill')

```
✔ **Useful when you expect future values to be similar.**

---

### **3.6 Filling NaNs with Different Values for Different Columns**
```python
values = {'A': df['A'].mean(), 'B': df['B'].median(), 'C': 0}
df_custom = df.fillna(value=values)

```
✔ **Useful when different columns need different strategies for handling NaNs.**

---

### **3.7 Limiting the Number of NaNs Filled**
```python
df_limited = df.fillna(method='ffill', limit=1)
print("Limited Forward Fill:\n", df_limited)
```
✔ **Prevents overfilling in case of consecutive NaNs.**

```
✔ **Works the same way as DataFrames but for a single column of data.**

---

## **5. Best Practices and Considerations**
### **✔ When to Use Which Filling Method?**
| **Scenario** | **Best Method** |
|--------------|----------------|
| Missing values are rare and should be ignored | `dropna()` |
| Missing values are small in number and should not affect analysis | Replace with `0` or `Unknown` |
| Dataset contains continuous numerical data | Use `fillna(df.mean())` |
| Dataset has categorical values | Use `fillna(df.mode()[0])` |
| Time series or sequential data | Use `fillna(method='ffill')` or `fillna(method='bfill')` |

### **✔ Things to Watch Out For**
1. **Don't blindly replace NaNs** without understanding their significance.
2. **Avoid overfilling** in time-series data—use the `limit` parameter.
3. **Mean/Median/Mode Filling Assumptions**  
   - Use mean if data is normally distributed.  
   - Use median if data is skewed.  
   - Use mode for categorical values.

---

## **6. Summary Table**
| Parameter | Description | Example |
|-----------|-------------|---------|
| `value` | Replace NaNs with a fixed value | `df.fillna(0)` |
| `method` | Fill using forward (`ffill`) or backward (`bfill`) fill | `df.fillna(method='ffill')` |
| `axis` | Fill along rows (`0`) or columns (`1`) | `df.fillna(axis=1, method='bfill')` |
| `inplace` | Modify DataFrame in place (`True`) or return a new one (`False`) | `df.fillna(0, inplace=True)` |
| `limit` | Maximum number of NaNs to fill | `df.fillna(method='ffill', limit=1)` |

2  2.0  3.0  2.5
3  4.0  4.0  4.0
```
✅ **Each NaN is replaced with the median of its respective column.**  
- Median of column `A`: `median([1, 2, 4]) = 2`
- Median of column `B`: `median([2, 3, 4]) = 3`
- Median of column `C`: `median([1, 4]) = 2.5`

---

## **4️⃣ Forward Fill (`ffill`)**
### **Code:**
```python
df_ffill = df.fillna(method='ffill')
print("\nForward Fill:\n", df_ffill)
```

### **Output:**
```
Forward Fill:
      A    B    C
0  1.0  NaN  1.0
1  2.0  2.0  1.0
2  2.0  3.0  1.0
3  4.0  4.0  4.0
```
✅ **Each NaN is replaced with the previous row's value.**  
- `A[2]` takes the value from `A[1]` (2.0)  
- `B[0]` remains `NaN` because there's no previous value.  
- `C[1]` and `C[2]` take the previous available value `C[0]` (1.0).  

---

## **5️⃣ Backward Fill (`bfill`)**
### **Code:**
```python
df_bfill = df.fillna(method='bfill')
print("\nBackward Fill:\n", df_bfill)
```

### **Output:**
```
Backward Fill:
      A    B    C
0  1.0  2.0  1.0
1  2.0  2.0  4.0
2  4.0  3.0  4.0
3  4.0  4.0  4.0
```
✅ **Each NaN is replaced with the next row's value.**  
- `A[2]` takes the value from `A[3]` (4.0)  
- `B[0]` takes the value from `B[1]` (2.0)  
- `C[1]` and `C[2]` take the next available value `C[3]` (4.0).

### apply(series + dataframe)


**Description** : Apply a function along an axis of the DataFrame.



````python
# phle apply ko series me use karte hia
temp = pd.Series([10,20,30,40,50])
temp
````

````python
def sigmoid(value):
    return 1/1+np.exp(-value)

temp.apply(sigmoid)
````

````python
# Now apply ko DataFrame ke upar use karte hai

points_df = pd.DataFrame(
    {
        '1st point':[(3,4),(-6,5),(0,0),(-10,1),(4,5)],
        '2nd point':[(-3,4),(0,0),(2,2),(10,10),(1,1)]
    }
)

points_df
````

````python
def euclidean(row):
    point_A = row['1st point']
    point_B = row['2nd point']
    
    return ((point_A[0] - point_B[0]) ** 2 + (point_A[1] - point_B[1])**2) ** 0.5
````

````python
points_df['distance'] = points_df.apply(euclidean, axis = 1)

points_df
````

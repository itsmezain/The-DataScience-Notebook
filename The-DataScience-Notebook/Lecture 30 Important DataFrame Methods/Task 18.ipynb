### `Q-6:` Make a dataframe of each team in IPL with details like - Team Name, Matches Played, Win%, Home Win%, Away Win%.
Show sorted dataframe on Win%

Replace old team name as new name before performing any tasks.
```
Delhi Daredevils ->Delhi Capitals
Kings XI Punjab -> Punjab Kings
Rising Pune Supergiants -> Rising Pune Supergiant
```

Note: Team1 represents Home team. Exclude not result matches.


````python
print(ipl_matches.shape)
ipl_matches.head(3)
````

````python
# code here

ipl_matches.replace(
    to_replace={
        'Delhi Daredevils' : 'Delhi Capitals',
        'Kings XI Punjab' : 'Punjab Kings',
        'Rising Pune Supergiants' : 'Rising Pune Supergiant'
    }, inplace= True
)
````

````python
ipl_matches_with_wining_team = ipl_matches[~ipl_matches["WinningTeam"].isna()]

teams = ipl_matches_with_wining_team.Team1.unique()

data = []

for team in teams:
    
    match_count = ipl_matches_with_wining_team[(ipl_matches_with_wining_team['Team1'] == team ) |  (ipl_matches_with_wining_team['Team2'] == team)].shape[0]
    
    winning_count = ipl_matches_with_wining_team[ipl_matches_with_wining_team["WinningTeam"] == team].shape[0]
    
    winning_percentage = round((winning_count / match_count) * 100, 2)

    # Home matches calculation
    
    home_matches_count = ipl_matches_with_wining_team[(ipl_matches_with_wining_team['Team1'] == team )].shape[0]

    home_winning_count = ipl_matches_with_wining_team[(ipl_matches_with_wining_team["WinningTeam"] == team) & (ipl_matches_with_wining_team['Team1'] == team)].shape[0]
    
    home_winning_percentage = round((home_winning_count / home_matches_count) * 100, 2)
    
    
    # Away matches calculation
    
    away_matches_count = ipl_matches_with_wining_team[(ipl_matches_with_wining_team['Team2'] == team )].shape[0]

    away_winning_count = ipl_matches_with_wining_team[(ipl_matches_with_wining_team["WinningTeam"] == team) & (ipl_matches_with_wining_team['Team2'] == team)].shape[0]
    
    away_winning_percentage = round((away_winning_count / away_matches_count) * 100, 2)
    
    
    data.append(
        [
            team,
            match_count,
            winning_percentage,
            home_winning_percentage,
            away_winning_percentage
        ]
    )
    
team_performance_df = pd.DataFrame(
    data = data,
    columns = ['Team Name', 'Matches Played', 'Win %', 'Home Win %', 'Away Win %']
).sort_values('Win %', ascending= False)

team_performance_df
````

### `Q-9:` IPL Point Table

Make a function `point_table` which take `season` as parameter and show points table in non-ascendng order of points and in ascending order of team name.

For winning - 2 Ponits;
For loosing - 0 Point
For not result both team gets 1 points.

Make dataframe which will have
`TeamName` `MatchesPlayed` `MatchesWon` `NoResult` `Points`
make `TeamName` as index.

```
season parametr should be one of these->
['2022', '2021', '2020/21', '2019', '2018', '2017', '2016', '2015', '2014', '2013', '2012', '2011', '2009/10', '2009', '2007/08']
```


Output of two Top 2 in season 2022
```
TeamName    MatchesPlayed	MatchesWon	NoResult	Points

Gujarat Titans	    16	   12	       0	     24
Rajasthan Royals	  17	   10	       0	     20

```

````python
ipl_matches.head(1)
````

````python
# code here

def point_table(season = "2007/08"):

    matches_in_season = ipl_matches[ipl_matches['Season'] == season]
    
    teams_list = ipl_matches[ipl_matches['Season'] == season].Team1.unique()
    
    data_rows = []

    for team in teams_list:
        matches_played = matches_in_season[(matches_in_season["Team1"] == team) | (matches_in_season["Team2"] == team)].shape[0]
        
        matches_won = matches_in_season[matches_in_season["WinningTeam"] == team].shape[0]
        
        no_result_matches = matches_in_season[ matches_in_season["WinningTeam"].isna() & ((matches_in_season["Team1"] == team) | (matches_in_season["Team2"] == team))]. shape[0]
        
        
        points = (matches_won * 2) + (no_result_matches * 1)
        
        data_rows.append({
            "TeamName": team,
            "MatchesPlayed": matches_played,
            "MatchesWon": matches_won,
            "NoResult": no_result_matches,
            "Points": points
        })
        
        point_table_df = pd.DataFrame(data_rows).set_index("TeamName")

        point_table_df.sort_values(by=["Points", "TeamName"], ascending=[False, True], inplace=True)
        
        
    return point_table_df


point_table("2022")

````

### `Q-10:` IPL Point Table cont.
Extend the above IPL Point Table with an extra column as `SeasonPosition`

Team below top 4 after sorting on `Points` and then on `TeamName` Will have same `SeasonPosition` as there rank. use rank function.

Teams in Top four will have `SeasonPosition` as:
```
    'Winner' - Team won final
    'Runner' - Team lost Final
    3 - Losing Team in Qualifier2
    4 - Losing Team in Eliminator
```

For changing value of pariticular cell use `df.at[row_index, col_label] = value`

Output of two Top 2 in season 2022. Your result should have all teams.
```
TeamName    MatchesPlayed	MatchesWon	NoResult	Points   SeasonPosition

Gujarat Titans	    16	   12	       0	     24         Winner
Rajasthan Royals	  17	   10	       0	     20         Runner

```

Note: If you try to chnage value of view of any dataframe a warnig will be shown. To avoid it, make a copy of the dataframe you want to change in by `df.copy()`

````python
ipl_matches.head()
````

````python
def point_table_extension(season="2022"):
    # Filter matches for the given season
    matches_in_season = ipl_matches[ipl_matches['Season'] == season]
    
    # Get the base point table (make a copy to avoid view warnings)
    table = point_table(season).copy()
    
    # Initially assign SeasonPosition as numeric rank based on Points (descending order)
    table['SeasonPosition'] = table["Points"].rank(method="first", ascending=False).astype("object")
    
    # --- Process the Final Match ---
    final_match = matches_in_season[matches_in_season["MatchNumber"] == "Final"]
    if not final_match.empty:
        final_match = final_match.iloc[0]
        winning_team = final_match["WinningTeam"]
        table.at[winning_team, "SeasonPosition"] = "Winner"
        
        # Determine the runner-up (the team that did not win)
        team1 = final_match["Team1"]
        team2 = final_match["Team2"]
        runner = team1 if team1 != winning_team else team2
        table.at[runner, "SeasonPosition"] = "Runner"
    
    # --- Process Qualifier 2: Losing team gets "Third" ---
    qualifier2_match = matches_in_season[matches_in_season["MatchNumber"].str.contains("Qualifier 2", na=False)]
    if not qualifier2_match.empty:
        qualifier2_match = qualifier2_match.iloc[0]
        qualifier2_winner = qualifier2_match["WinningTeam"]
        qualifier2_team1 = qualifier2_match["Team1"]
        qualifier2_team2 = qualifier2_match["Team2"]
        qualifier2_loser = qualifier2_team1 if qualifier2_team1 != qualifier2_winner else qualifier2_team2
        table.at[qualifier2_loser, "SeasonPosition"] = "Third"
    
    # --- Process Eliminator: Losing team gets "Fourth" ---
    eliminator_match = matches_in_season[matches_in_season["MatchNumber"] == "Eliminator"]
    if not eliminator_match.empty:
        eliminator_match = eliminator_match.iloc[0]
        eliminator_winner = eliminator_match["WinningTeam"]
        eliminator_team1 = eliminator_match["Team1"]
        eliminator_team2 = eliminator_match["Team2"]
        eliminator_loser = eliminator_team1 if eliminator_team1 != eliminator_winner else eliminator_team2
        table.at[eliminator_loser, "SeasonPosition"] = "Fourth"
    
    return table

# Example usage:
pt_extended = point_table_extension("2022")
pt_extended
````

```python
from sqlalchemy import create_engine, text
import os
from dotenv import load_dotenv

load_dotenv()


engine_for_olap = create_engine(
    f"mysql+pymysql://{os.getenv('DB_USER')}:{os.getenv('DB_PASSWORD')}@{os.getenv('DB_HOST')}:{os.getenv('DB_PORT')}"
)

with engine_for_olap.connect() as conn:
    conn.execute(text("CREATE DATABASE IF NOT EXISTS dream11_olap"))
    conn.commit()

print("Database created")
```

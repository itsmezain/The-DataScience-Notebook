<span style="font-size: 18px;"> <b>Question: Find the order name and corresponding user name

```python
%%sql

# As you see that order_details cannot be join with user_id directly as it don't have any common columns. Hence we are gonna first join order_details and order and then user

SELECT *
FROM order_details t1
JOIN orders t2
ON t1.order_id = t2.order_id
JOIN users t3
ON t2.user_id = t3.user_id
```

<span style="font-size: 18px;"> <b>Question: Find order_id, amount, profit, name, and city by joining order_details, users and orders 

```python
%%sql

SELECT 
    t1.order_id, 
    t1.amount, 
    t1.profit, 
    t3.name, 
    t3.city
FROM order_details t1
JOIN orders t2
ON t1.order_id = t2.order_id
JOIN users t3
ON t2.user_id = t3.user_id
```

# <center> Practice Questions


<span style="font-size: 18px;"> <b>Question: Find all profitable orders

```python
%%sql

SELECT
    t1.order_id,
    SUM(t2.profit)
FROM orders t1
JOIN order_details t2
ON t1.order_id = t2.order_id
GROUP BY t2.order_id
HAVING SUM(t2.profit) > 0;
```

<span style="font-size: 18px;"> <b>Question: Find the customer who has placed the max number of orders 

```python
%%sql

SELECT
    t2.name,
    t2.user_id,
    COUNT(*)        AS 'No_of_order'
FROM orders t1
JOIN users t2
ON t1.user_id = t2.user_id
GROUP BY t1.user_id, t2.name, t2.user_id # here t2.name is added bcz in the select keyword we can only write the columns name whose is there in the group by or on which aggregation operation is perform
ORDER BY No_of_order DESC
LIMIT 1;
```

<span style="font-size: 18px;"> <b>Question: Which is the most profitable category

```python
%%sql

SELECT
    t2.vertical     AS 'Category',
    SUM(profit)     AS 'total_profit'
FROM order_details t1
JOIN category t2
ON t1.category_id = t2.category_id
GROUP BY vertical
ORDER BY total_profit DESC
LIMIT 1;
```

<span style="font-size: 18px;"> <b>Question: Which is the most profitable state 

```python
%%sql

SELECT
    state,
    SUM(profit)         AS 'total_profit'
FROM order_details t1
JOIN orders t2
ON t1.order_id = t2.order_id
JOIN users t3
ON t2.user_id = t3.user_id
GROUP BY state
ORDER BY total_profit DESC
```

<span style="font-size: 18px;"> <b>Question: Find all categories with profits higher than 4000 

```python
%%sql

SELECT
    vertical,
    SUM(profit)         AS 'total_profit'
FROM order_details t1
JOIN category t2
ON t1.category_id = t2.category_id
GROUP BY vertical
HAVING total_profit > 4000
```

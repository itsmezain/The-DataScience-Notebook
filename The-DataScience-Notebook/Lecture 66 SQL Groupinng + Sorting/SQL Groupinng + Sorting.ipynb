# <center> Sorting Data

<span style="font-size: 18px;"> <b>Question: Find top 5 samsung phones with the biggest screen size.

```python
%%sql

SELECT Product_Name, display_size_inches 
FROM smartphone.mobiles
WHERE brand_name = 'samsung'
ORDER BY display_size_inches  DESC
LIMIT 5;
```

<span style="font-size: 18px;"> <b>Question: Sort the phones according to the number of rear cameras they have.

```python
%%sql
SELECT
    Product_Name      AS 'Phone Model',
    brand_name        AS 'Brand',
    rear_camera_count AS 'No. of Cameras'
FROM mobiles
ORDER BY rear_camera_count DESC;
```

<span style="font-size: 18px;"> <b>Question: Sort data on the basis of ppi in descending order

```python
%%sql

SELECT 
    Product_Name    AS 'Phone Model',
    ROUND(
        SQRT(POW(resolution_width, 2) + POW(resolution_height,2)) / display_size_inches, 2
    ) AS PPI
FROM mobiles
ORDER BY PPI DESC;
```

<span style="font-size: 18px;"> <b>Question: Find the phones with 2nd largest battery

```python
%%sql

SELECT 
    Product_Name    AS 'Phone Model',
    battery_capacity_mah AS 'Battery in MAH'
FROM mobiles
ORDER BY battery_capacity_mah DESC
LIMIT 1,1; # OFFSET by, number of rows
```

<span style="font-size: 18px;"> <b>Question: Find the phones with 3rd,4th and 5th smallest battery

```python
%%sql

SELECT
    Product_Name        AS 'Phone Model',
    battery_capacity_mah AS 'Battery in MAH'
FROM mobiles
ORDER BY battery_capacity_mah ASC
LIMIT 2,3;
```

<span style="font-size: 18px;"> <b>Question: Find the all the phones with 7nd largest battery i.e. 7000 

```python
%%sql
SELECT Product_Name, brand_name, battery_capacity_mah
FROM mobiles
WHERE battery_capacity_mah = (
    SELECT DISTINCT battery_capacity_mah
    FROM mobiles
    ORDER BY battery_capacity_mah DESC
    LIMIT 1 OFFSET 6      -- OFFSET 0=1st, 1=2nd, 6=7th
);
```

<span style="font-size: 18px;"> <b>Question: Find the name and rating of the worst spec_score apple phone.

```python
%%sql

SELECT
    Product_Name        AS 'Phone Name',
    Spec_Score          AS 'Spec Score'
FROM mobiles
WHERE brand_name = 'apple'
ORDER BY Spec_Score ASC
LIMIT 1;
```

<span style="font-size: 18px;"> <b>Question: Sort the data based on the brand_name and price

```python
%%sql

SELECT 
    Product_Name        AS 'Phone Name',
    brand_name          AS 'Brand Name',
    Price               AS 'Price'
FROM mobiles
ORDER BY brand_name ASC, Price DESC;
```

<span style="font-size: 18px;"> <b>Question: Sort the data by the brand_name and its spec_score.

```python
%%sql

SELECT 
    Product_Name        AS 'Phone Name',
    brand_name          AS 'Brand Name',
    Spec_Score          AS 'Spec Score'
FROM mobiles
ORDER BY brand_name ASC, Spec_Score DESC;
```

# <center> Grouping Data

<span style="font-size: 18px;"> <b>Question: Group smartphones by brand and get the count, average price, max spec_score, avg screen size and avg battery capacity

```python
%%sql

SELECT
    brand_name,
    COUNT(Product_name)     As 'No. of Phones',
    ROUND(AVG(price))       AS 'Avg Price',
    MAX(Spec_score)         AS 'Max Spec Score',
    ROUND(AVG(display_size_inches), 2) AS 'Avg Screen size',
    ROUND(AVG(battery_capacity_mah)) AS 'Avg battery capacity'
FROM mobiles
GROUP BY brand_name;
```

<span style="font-size: 18px;"> <b>Question: Group smartphones by whether they have an NFC and get the average price and rating

```python
%%sql

SELECT
    has_nfc,
    Round(AVG(Price))          AS 'Avg Price',
    Round(AVG(Spec_score))     AS 'AVG Spec Score'
FROM mobiles
GROUP BY has_nfc;
```

```python
%%sql

SELECT
    has_5g,
    Round(AVG(Price))          AS 'Avg Price',
    Round(AVG(Spec_score))     AS 'AVG Spec Score'
FROM mobiles
GROUP BY has_5g;
```

<span style="font-size: 18px;"> <b>Question: Group smartphones by the extended memory available and get the average price

```python
%%sql

SELECT
    card_supported,
    ROUND(AVG(Price))       AS 'Avg Price'
FROM mobiles
GROUP BY card_supported;
```

<span style="font-size: 18px;"> <b>Question: Group smartphones by the brand and processor brand and get the count of models and the average primary camera resolution(rear)

```python
%%sql

SELECT
    brand_name,
    processor_brand,
    Count(price),
    ROUND(AVG(rear_camera_mp))      AS 'Avg Primary Camera M.P.'
FROM mobiles
GROUP BY brand_name,processor_brand;
```

<span style="font-size: 18px;"> <b>Question: Find the top 5 most costly phones brand.

```python
%%sql

SELECT
    brand_name,
    ROUND(AVG(Price))          AS 'Avg_price'
FROM mobiles
GROUP BY brand_name
ORDER BY Avg_price DESC
Limit 5;
```

<span style="font-size: 18px;"> <b>Question: Which brand makes the smallest screen smartphones

```python
%%sql

SELECT
    brand_name,
    AVG(display_size_inches)        AS 'Avg_screen_size'
FROM mobiles
GROUP BY brand_name
ORDER BY Avg_screen_size ASC
LIMIT 5;
```

<span style="font-size: 18px;"> <b>Question: Group smartphones by the brand and find the brand with the highest number of phones that have both NFC and an IR blaster

```python
%%sql

SELECT
    brand_name,
    COUNT(*)        AS 'No_of_phones_with_NFC_and_IR_blaster'
FROM mobiles
WHERE has_nfc = 1 AND has_ir_blaster = 1
GROUP BY brand_name
ORDER BY No_of_phones_with_NFC_and_IR_blaster DESC;   
```

<span style="font-size: 18px;"> <b>Question: Find all samsung 5g enables smartphones and find out the avg price for NFC and Non-NFC phones

```python
%%sql

SELECT
    has_nfc,
    Round(AVG(Price))       AS 'avg_price'
FROM mobiles
WHERE brand_name = 'samsung' And has_5g = 1
GROUP BY has_nfc
```

<span style="font-size: 18px;"> <b>Question: Find the phones name and price of the costliest phone

```python
%%sql

SELECT
    Product_Name,
    Price
FROM mobiles
ORDER BY Price DESC
Limit 1;
```

# <center>Having Clause

Select ke liye jo kaam WHERE krta hai,  
Group By ke liye woo kaam HAVING krta hai.

<span style="font-size: 18px;"> <b>Question: Find the avg price of the phones brand which have more than 20 phones.

```python
%%sql

SELECT
    brand_name,
    COUNT(*)        AS 'count',              
    ROUND(AVG(Price))      AS 'avg_price'
FROM mobiles
GROUP BY brand_name
HAVING COUNT(*) > 20
ORDER BY avg_price DESC;
```

<span style="font-size: 18px;"> <b>Question: Find the spec_score of the smartphones brand which have more than 20 phones

```python
%%sql

SELECT
    brand_name,
    AVG(Spec_Score)     AS 'avg_spec_score'
FROM mobiles
GROUP BY brand_name
HAVING COUNT(*) > 20
ORDER BY avg_spec_score DESC;
```

<span style="font-size: 18px;"> <b>Question: Find the top 3 brands with the highest avg RAM that have a refresh rate of at least 90 HZ and fast charging available and dont consider brand which less than 10 phones

```python
%%sql

SELECT
    brand_name,
    ROUND(AVG(ram_gb))     AS 'avg_ram'
FROM mobiles
WHERE display_frequency_hz >= 90 AND fast_charging_available = 1
GROUP BY brand_name
HAVING COUNT(*) > 10
ORDER BY avg_ram DESC
LIMIT 3;
```

<span style="font-size: 18px;"> <b>Question: Find the avg price of all the phone brands with avg spec_score > 75 and num_phones more than 10 among all 5g enabled phones

```python
%%sql

SELECT
    brand_name,
    AVG(Price)
FROM mobiles
WHERE has_5g = 1
GROUP BY brand_name
HAVING AVG(Spec_Score) > 75 AND COUNT(*) > 10;
```

<span style="font-size: 18px;"> <b>Question: Find the top 5 batsmen runs scored in IPL

```python
%%sql

SELECT
    batter,
    SUM(batsman_run)        AS 'runs_scored'
FROM ipl_data
GROUP BY batter
ORDER BY runs_scored DESC
LIMIT 5;
```

<span style="font-size: 18px;"> <b>Find the 2nd highest 6 hitters in IPL

```python
%%sql

SELECT 
    batter,
    COUNT(*)          AS 'six_hitted'
FROM ipl_data
WHERE batsman_run = 6
GROUP BY batter
ORDER BY six_hitted DESC
LIMIT 1,1;
```

<span style="font-size: 18px;"> <b>Question: Find the batsmen name who scored century with each century score in IPL

```python
%%sql

SELECT
    batter,
    ID,
    SUM(batsman_run)        AS 'runs_scored'
FROM ipl_data
GROUP BY batter, ID
HAVING runs_scored >= 100
ORDER BY batter ASC;
```

<span style="font-size: 18px;"> <b>Find the top 5 batsmen with the highest strike rate who have played a min of 1000 balls

```python
%%sql

SELECT
    batter,
    SUM(batsman_run)        AS 'runs_scored',
    COUNT(*)                AS 'balls_played',
    (SUM(batsman_run) / COUNT(*)) * 100 AS 'strike_rate'
FROM ipl_data
WHERE extra_type != 'wides' OR extra_type IS NULL
GROUP BY batter
HAVING balls_played >= 1000
ORDER BY strike_rate DESC
LIMIT 5;
```

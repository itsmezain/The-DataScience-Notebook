> method 2

1- Add a Temporary ID
```sql
ALTER TABLE laptopData ADD COLUMN temp_id INT AUTO_INCREMENT PRIMARY KEY;
```  

2-  Then Delete
```sql
DELETE t1 FROM laptopData t1
INNER JOIN (
    SELECT temp_id,
           ROW_NUMBER() OVER (
               PARTITION BY Company, TypeName, Inches, ScreenResolution, Cpu, Ram, Memory, Gpu, OpSys, Weight, Price
               ORDER BY temp_id
           ) AS row_num
    FROM laptopData
) t2 ON t1.temp_id = t2.temp_id
WHERE t2.row_num > 1;
```

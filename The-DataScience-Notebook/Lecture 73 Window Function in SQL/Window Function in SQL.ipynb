# <center> RANK

- `RANK`: RANK function gives the ranking over the given group.

<span style="font-size: 18px;"> <b>`Question`: Find the ranks of the student on their marks.

```python
%%sql

SELECT
    *,
    RANK() OVER (ORDER BY marks	DESC)
FROM marks;
```

> As above you can see that the rank 1 is repeated 4 times but there is no rank 2,3,4.
>> This is because that how the RANK function works, if u dom't want this you can use the DENSE RANK.

<span style="font-size: 18px;"> <b>`Question`: Find the ranks of the student of each branch based on their marks.

```python
%%sql

SELECT
    *,
    RANK() OVER (PARTITION BY branch ORDER BY marks)
FROM marks;
```

# <center> DENSE RANK

- `RANK`: RANK function gives the ranking over the given group.

<span style="font-size: 18px;"> <b>`Question`: Find the ranks without missing any rank of the student based on their marks.

```python
%%sql

SELECT 
    *,
    RANK() OVER (PARTITION BY branch ORDER BY marks) as 'rank',
    DENSE_RANK() OVER (ORDER BY marks DESC) as 'dense_rank'
FROM marks;
```

<span style="font-size: 18px;"> <b>`Question`: In the zomato dataset, Find top 2 most paying customers.

```python
zomato_conn = 'mysql+pymysql://root:zain@localhost/zomato'
```

```python
%%sql $zomato_conn

-- Zain:
    
SELECT 
    user_id,
    Total_Spend
FROM (
    SELECT 
    *, 
    DENSE_RANK() OVER(ORDER BY Total_Spend DESC) as rnk
FROM (
    SELECT 
        *,
        SUM(amount) OVER(PARTITION BY user_id) AS 'Total_Spend'
    FROM orders
) t
) x
WHERE rnk = 1  OR rnk = 2
GROUP BY user_id, Total_Spend
    
```

<span style="font-size: 18px;"> <b>`Question`: Find top 2 most paying customers of each month.

```python
%%sql $zomato_conn

-- Step 1 + 2: Sum per user per month, then rank within each month

WITH monthly_spend AS (
    SELECT
        t1.user_id,
        t2.name,
        MONTH(date)      AS 'month_num',
        SUM(amount)      AS 'total_spend'
    FROM orders t1
    JOIN users t2 ON t1.user_id = t2.user_id
    GROUP BY t1.user_id, t2.name, MONTH(date)
),
ranked AS(
    SELECT 
        user_id,
        name,
        month_num,
        total_spend,
        DENSE_RANK() OVER (
            PARTITION BY month_num 
            ORDER BY total_spend
        )       AS rnk
    FROM monthly_spend
)

-- Step 3: Filter top 2
SELECT *
FROM ranked
WHERE rnk <= 2
```

```python
%%sql $zomato_conn

-- Nitish Sir method

SELECT * 
FROM (SELECT 
    MONTH(date),
    user_id,
    SUM(amount),
    RANK() OVER(PARTITION BY MONTH(date) ORDER BY SUM(amount) DESC) as rnk
FROM orders
GROUP BY MONTH(date), user_id
ORDER BY MONTH(date)
) t
WHERE t.rnk <= 2
```

## <center>Frame

> SUPER important concept for the window functions.  
THERE are 4 types of frames and based on which type you choice window function gonna give answer differently.  

• ROWS BETWEEN UNBOUNDED PRECEDING AND CURRENT ROW `Default`  
• ROWS BETWEEN 1 PRECEDING AND 1 FOLLOWING  
• ROWS BETWEEN UNBOUNDED PRECEDING AND UNBOUNDED FOLLOWING  
• ROWS BETWEEN 3 PRECEDING AND 2 FOLLOWING  

> FRAME is a dynamic subset of data from which the window function evaluates result for every individual row.  

FRAME can choice subset of data in the following ways:
- All rows included in the frame
- Rows are chosen in between from the some rows, jaise 1 row se current row tkk data se last row nikla, jaise current row me aur current row ke 2 upr se last row nikalna.  

- example: jaise ke last row nikla huva aur ham ne ROW BETWEEN UNBOUNDED PRECEDING AND CURRENT ROW frame choice kiya hai too.  

First baar me last row last wale ho jayege
SECOND baar me last row 1 & 2 row me se last wale means 2nd row ho jayegi.
Similarly, Third baar me last row 1,2 and 3 me se last wale ho jayegi means 3rd wale.

<span style="font-size: 18px;"> <b>`Question`: Find the lowest marks student using all rows in the frame.

```python
%%sql $lecture_conn

-- 
SELECT
    *,
    LAST_VALUE(marks) OVER(ORDER BY marks DESC
                            ROWS BETWEEN UNBOUNDED PRECEDING AND UNBOUNDED FOLLOWING)
FROM marks;
```

<span style="font-size: 18px;"> <b>`Question`: Find the lowest marks per branch using all rows in the frame.

```python
%%sql $lecture_conn

SELECT 
    *,
    LAST_VALUE(marks) OVER(
            PARTITION BY branch 
            ORDER BY marks DESC
            ROWS BETWEEN UNBOUNDED PRECEDING AND UNBOUNDED FOLLOWING)
FROM marks;
```

<span style="font-size: 18px;"> <b>`Question`: Find the name of the topper per branch using all rows in the frame.

# <center> Shortened WINDOW function

<span style="font-size: 18px;"> <b>`Question`: Find the branch lowest marks student name, branch name.

```python
%%sql $lecture_conn

SELECT 
    name,
    branch,
    marks
FROM (
    SELECT 
        *,
        LAST_VALUE(marks) OVER w    AS 'topper_marks',
        LAST_VALUE(student_id) OVER w   AS 'topper_user_id'
    FROM marks
    WINDOW w AS (
        PARTITION BY branch
        ORDER BY marks DESC 
        ROWS BETWEEN UNBOUNDED PRECEDING AND UNBOUNDED FOLLOWING)
) t
WHERE t.marks = t.topper_marks AND 
    t.student_id = t.topper_user_id;
```

<span style="font-size: 18px;"> <b>`Question`: Find the branch toppers name, branch name.

```python
%%sql $lecture_conn

SELECT 
    name,
    branch,
    marks
FROM (
    SELECT 
    *,
    FIRST_VALUE(marks) OVER(PARTITION BY branch
                    ORDER BY marks DESC)   AS 'topper_marks',
    FIRST_VALUE(student_id) OVER(PARTITION BY branch
                    ORDER BY marks DESC)   AS 'topper_user_id'
FROM marks
) t
WHERE t.marks = t.topper_marks AND 
    t.student_id = t.topper_user_id
```

<span style="font-size: 18px;"> <b>`Question`: Find the branch lowest marks student name, branch name.

```python
%%sql $lecture_conn

SELECT 
    name,
    branch,
    marks
FROM (
    SELECT 
    *,
    LAST_VALUE(marks) OVER(
        PARTITION BY branch
        ORDER BY marks DESC 
        ROWS BETWEEN UNBOUNDED PRECEDING AND UNBOUNDED FOLLOWING)   AS 'topper_marks',
    LAST_VALUE(student_id) OVER(
        PARTITION BY branch
        ORDER BY marks DESC 
        ROWS BETWEEN UNBOUNDED PRECEDING AND UNBOUNDED FOLLOWING)   AS 'topper_user_id'
FROM marks
) t
WHERE t.marks = t.topper_marks AND 
    t.student_id = t.topper_user_id
```

# <center> LAG

In MySQL, LAG() is a window function used to access data from a previous row in the same result set.

<span style="font-size: 18px;"> <b>`Question`: Find the marks student with 1 lag

```python
%%sql $lecture_conn

SELECT
    *,
    LAG(marks) OVER(ORDER BY student_id)
FROM marks;
```

<span style="font-size: 18px;"> <b>`Question`: Find the marks student with 2 lag

```python
%%sql $lecture_conn

SELECT
    *,
    LAG(marks, 2) OVER(ORDER BY student_id)
FROM marks;
```

<span style="font-size: 18px;"> <b>`Question`: Find the MoM revenue growth of Zomato.

```python
%%sql $zomato_conn

Select
    MONTH(date),
    SUM(amount),
    ((SUM(amount) - LAG(SUM(amount)) OVER(ORDER BY MONTH(date))) /  LAG(SUM(amount)) OVER(ORDER BY MONTH(date)) )* 100 AS 'Growth_MoM'
FROM orders
GROUP BY MONTH(date) 
```
